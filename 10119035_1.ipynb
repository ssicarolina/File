## 1

Buat program Python untuk menghitung ekspresi berikut, **tanpa memakai tanda kurung**. Tampilkan hasilnya pada layar komputer.

$$\frac{a+bc}{ac + \frac{c}{a^2+b^3}} \quad \text{a, b, c integer}$$

Tidak boleh ada variabel tambahan lain selain `a`, `b`, `c`. Ujilah program tersebut dengan nilai `a = 3`, `b = 4`, dan `c = 5` yang dimasukkan dengan perintah input.

````
# Program penghitung-ekspresi
# Menghitung nilai ekspresi
#
#                        a + bc
#                   ----------------
#                             c
#                    ac + ---------
#                          a^2+b^3
#
# dan menampilkannya pada output
#
# Variable:
#   a,b,c   variabel pada ekspresi
#

a = int(input("Masukkan nilai a: "))
b = int(input("Masukkan nilai b: "))
c = int(input("Masukkan nilai c: "))

c = a**2/c + b**3/c + a*b + b**4/a
a = a**2/c + b**3/c + 1/a/c

print("Nilai ekspresi adalah", 1/a)
````


````
# Program kura-kura-kancil
# Menghitung waktu yang diperlukan kancil dengan kecepatan V2 cm/detik
# untuk menyusul kura-kura dengan kecepatan V1 cm/detik
# apabila kancil mulai bergerak k detik lebih akhir daripada kura-kura
#
# Variable:
#   V1_1, V1_2, V1_3, V1_4 kecepatan kura-kura (cm/detik)
#   V2_1, V2_2, V2_3, V2_4 kecepatan kancil (cm/detik)
#   k_1, k_2, k_3, k_4 selisih waktu mulai kancil dan kura-kura (detik)
#   T_1, T_2, T_3, T_4 waktu yang diperlukan kancil untuk menyusul kura-kura
#   D_1, D_2, D_3, D_4 jarak kancil menyusul kura-kura dari garis start
#

print("Test case 1:")
V1_1 = float(input("Masukkan kecepatan kura-kura (cm/detik): "))
V2_1 = float(input("Masukkan kecepatan kancil (cm/detik): "))
k_1  = float(input("Masukkan delay mulainya kancil (detik): "))
print("Test case 2:")
V1_2 = float(input("Masukkan kecepatan kura-kura (cm/detik): "))
V2_2 = float(input("Masukkan kecepatan kancil (cm/detik): "))
k_2  = float(input("Masukkan delay mulainya kancil (detik): "))
print("Test case 3:")
V1_3 = float(input("Masukkan kecepatan kura-kura (cm/detik): "))
V2_3 = float(input("Masukkan kecepatan kancil (cm/detik): "))
k_3  = float(input("Masukkan delay mulainya kancil (detik): "))
print("Test case 4:")
V1_4 = float(input("Masukkan kecepatan kura-kura (cm/detik): "))
V2_4 = float(input("Masukkan kecepatan kancil (cm/detik): "))
k_4  = float(input("Masukkan delay mulainya kancil (detik): "))

T_1 = V1_1*k_1/(V2_1-V1_1)
D_1 = V2_1*T_1

T_2 = V1_2*k_2/(V2_2-V1_2)
D_2 = V2_2*T_2

# T_3 = V1_3*k_3/(V2_3-V1_3)
# Seharusnya rumus T_3 seperti di atas, tetapi karena V1 sama dengan V2,
# rumus tersebut menghasilkan error division by zero.
# Di skenario soal, karena kancil lari secepat kura-kura dan kura-kura
# mulai lebih dahulu, kancil tidak akan bisa menyusul kura-kura.
T_3 = 0 # sebagai placeholder
D_3 = V2_3*T_3

T_4 = V1_4*k_4/(V2_4-V1_4)
D_4 = V2_4*T_4
# T_4 dan D_4 negatif karena V2 lebih kecil daripada V1.
# Di skenario soal, karena kancil lari lebih lambat daripada kura-kura,
# kancil tidak akan bisa menyusul kura-kura.

print("_____________________________________________________________________")
print("| V kura-kura |  V kancil   |   k   | Waktu bertemu | Jarak bertemu |")
print("|_____________|_____________|_______|_______________|_______________|")
print("| %11.1f |"%V1_1, "%11.1f |"%V2_1, "%5.1f |"%k_1,
      "%13.2f |"%T_1, "%13.2f |"%D_1)
print("| %11.1f |"%V1_2, "%11.1f |"%V2_2, "%5.1f |"%k_2,
      "%13.2f |"%T_2, "%13.2f |"%D_2)
print("| %11.1f |"%V1_3, "%11.1f |"%V2_3, "%5.1f |"%k_3,
      "%13.2f |"%T_3, "%13.2f |"%D_3)
print("| %11.1f |"%V1_4, "%11.1f |"%V2_4, "%5.1f |"%k_4,
      "%13.2f |"%T_4, "%13.2f |"%D_4)
print("|_____________|_____________|_______|_______________|_______________|")
````
