```python
# Nomor 1a

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square
from sympy import fourier_series, pi
from sympy.abc import t
from sympy import symbols
from sympy import *

f = (pi+t)*Heaviside(t+pi) - (pi+t)*Heaviside(t) + (pi-t)*Heaviside(t) - (pi-t)*Heaviside(t-pi) 
s = fourier_series(f, (t, -pi, pi))
s.truncate(40) # menampilkan 10 suku pertama deret fouriernya
deret = s.truncate(40)
print(deret)
p = plot(f, s.truncate(40), (t, -2*pi, 2*pi), show = False, legend = True)
p[0].line_color = 'red'
p[0].label = 'f(t)'
p[1].line_color = 'blue'
p[1].label = 'n = 40'
p.show()
```

```python
#1b
from sympy import fourier_series, pi
from sympy.abc import t
from sympy import *
f = (t)*(Heaviside(t+pi/2)-Heaviside(t-pi/2))+(pi-t)*((Heaviside(t-pi/2)-Heaviside(t-pi)))+(pi-t)*((Heaviside(t+pi)-Heaviside(t+pi/2)))
s = fourier_series(f, (t, -pi,pi))
s.truncate(40)

p = plot(f, s.truncate(40), (t, -3*pi, 3*pi), show = False, legend = True)
p[0].line_color = 'red'
p[0].label = 'f(t)'
p[1].line_color = 'blue'
p[1].label = 'n = 40'
p.show()
```

```python
# Nomor 1b

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square
from sympy import fourier_series, pi
from sympy.abc import t
from sympy import symbols
from sympy import *
f = (t)*Heaviside(t+(pi/2)) - (t)*Heaviside(t-(pi/2)) + (pi-t)*Heaviside(t-(pi/2)) - (pi-t)*Heaviside(t-(1.5*pi))
s = fourier_series(f, (t, -pi/2, 3*pi/2))
s.truncate(40) # menampilkan 10 suku pertama deret fouriernya

p = plot(f, s.truncate(40), (t, -3*pi, 3*pi), show = False, legend = True)
p[0].line_color = 'red'
p[0].label = 'f(t)'
p[1].line_color = 'blue'
p[1].label = 'n = 40'
p.show()
```
