```
import pdb
import random
N= random.randint(1,100)
n = random.random()
print(N)
x =0
y = 0
count = 0
while count <=N:
  if n<0.25:
    y = y+1       # move north east
    x = x+1
    n = random.random()
  if n>=0.25 and n<0.5:
    y = y-1       # move south east
    x = x+1
    n = random.random()
  if n>=0.5 and n<0.75:
    y = y+1       # move north west
    x = x-1
    n = random.random()
  if n>=0.75:
    y = y-1       # move south east
    x = x-1
    n = random.random()
  print("(%d,%d)"%(x,y))
  count = count + 1

print("squared distance = %d" %(x*x + y*y))
```

```
import pdb
import random
N= 3
n = random.random()
print(N)
x = 5
y = 8
count = 0
while count <=N:
  if n==0:
    n2 = random.random()
    if n2==0:
      x = x+1
    else:
      x = x-1
  else:
    n2 = random.random()
    if n2==0:
      y = y+1
    else:
      y = y-1
  print("(%d,%d)"%(x,y))
  count = count + 1

print("squared distance = %d" %(x*x + y*y))
```

```
# random walk latihan soal
#south east north west  
import pdb
import random  # this helps us generate random numbers
N = random.randint(1,100)  # number of steps
n = random.random()  # generate a random number
print(N)
x = 0
y = 0
count = 0 
while count <= N:
    if n < 0.10:
        y = y + 1           # move north
        n = random.random() # generate a new random number
    if n >= 0.10 and n < 0.15:
        y = y - 1           # move south
        n = random.random() # generate a new random number
    if n >= 0.15 and n < 0.55:
        x = x - 1           # move west
        n = random.random() # generate a new random number
    if n >= 0.55:
        x = x + 1           # move east
        n = random.random() # generate a new random number
    
    print("(%d,%d)" % (x,y)) 
    count = count + 1
print("squared distance = %d" % (x*x + y*y))
```
