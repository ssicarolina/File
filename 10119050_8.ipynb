# Catatan Animasi Heatmap

Pada modul kali ini, akan digunakan heatmap dari seaborn. Sehingga terdapat beberapa perbedaan kode dari modul animasi sebelumnya. Secara singkat, ada 3 modifikasi yang diperlukan.

Catatan:
```python
import matplotlib.pyplot as plt
import seaborn as sns
```
Sehingga, ```plt``` merujuk pada ```matplotlib.pyplot``` serta ```sns``` merujuk pada ```seaborn```

## 1. ```fig``` dan ```ax```


Karena kita tidak menggunakan plot dari matplotlib, maka terdapat sedikit modifikasi pada inisiasi ```fig``` dan ```ax```. 

- Plot menggunakan matplotlib:
```python
fix, ax = plt.subplots()
```

- Plot menggunakan heatmap seaborn:
```python
fig = plt.figure()
ax = sns.heatmap(...)
```

Secara singkat, kita akan meminta figure/window dari plot, serta isi plot dari seaborn heatmap.

## 2. ```set_data```

Pada modul animasi heatmap ini, ```.set_data()``` tidak dapat digunakan untuk mengupdate nilai yang akan di-plot. Sehingga, perbedaan mengubah data pada fungsi ```animate``` dari plot matplotlib dengan seaborn:

- Mengubah data jika plot biasa (matplotlib)
```python
def animate(i):
    line.set_data(...)
    dot.set_data(...)
```

- Mengubah data jika plot heatmap seaborn
```python
def animate(i):
    ax = sns.heatmap(...)
```

Secara singkat, kita hanya perlu mendefinisikan ulang variable ```ax``` menggunakn data yang baru.

## 3. ```init``` dan ```animate```

Karena ```set_data``` tidak dapat digunakan dan perlu untuk mendefinisikan kembali ```ax``` pada looping ```animate```, maka setiap kali melakukan iterasi pada fungsi ```animate``` plot harus 'dibersihkan' terlebih dahulu

Figure/window dapat 'dibersihkan' menggunakan fungsi ```plt.clf()```. Sehingga:

- Animasi matplotplib modul sebelumnya:
```python
def init():
    return None
```
```python
def animate(i):
    ...

    ...

```

- Animasi heatmap seaborn:
```python
def init():
    plt.clf()
    return None
```
```python
def animate(i):
    plt.clf()
    ...
    return None
```



## Tambahan: ```vmax``` dan ```vmin``` pada Heatmap

Agar warna tidak berubah selama animasi (karena perubahan nilai), setiap kali kita mendefinisikan ```ax = sns.heatmap(...)``` kita perlu untuk memberitahu program batas atas dan batas bawah nilai yang akan di plot.

Untuk melakukan itu, kita akan menambahkan parameter ```vmin``` dan ```vmax```.

Sehingga **setiap kali** kita mendefinisikan ```ax```, jika diinginkan batas bawah warna bernilai $a$ dan batas atas bernilai $b$:
```python
ax = sns.heatmap(data, vmin=a, vmax=b)
```

## Contoh

Pada contoh ini, akan dianimasikan langkah yang dilewati semut yang bergerak dari kiri ke kanan plot. Baris pergerakan diambil secara acak

````
# Library/fungsi yang wajib ada
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import seaborn as sns
import random

# Definisi 
x_size = 8
y_size = 12
data = [[0 for i in range(x_size)] for j in range(y_size)]
col_gerak = random.randint(0, y_size - 1)

# Menyiapkan figure, heatmap
fig = plt.figure()
ax = sns.heatmap(data, vmin=0, vmax=1)

# Fungsi init
def init():
    plt.clf()
    ax = sns.heatmap(data, vmin=0, vmax=1)
    return None

# Fungsi iterasi animasi
def animate(i):
    plt.clf()
    data[col_gerak][i - 1] = 1
    ax = sns.heatmap(data, vmin=0, vmax=1)
    return None

# Fungsi init
anim = animation.FuncAnimation(fig, animate, frames=range(1, x_size + 1, 1), blit=False,
                               interval=10, init_func=init)

# Kode tambahan jika dijalankan di colab/jupyter
from matplotlib import rc
from IPython.display import HTML
rc('animation', html='jshtml')
anim
````
