````
# Soal Latihan Nomor 1 
# mx'' + cx' + kx = F(t), x(0) = x_0, x'(0) = v_0
"""
Nomor 1
Untuk m = 1 F(t) = 0, x_0 = 1, dan x'(0) = 0, sketsa solusi persamaan diferensial tersebut 
untuk berbagai parameter berikut. Berikan interpretasi.
c = 5, k = 5,

Penyelesaian: 
a) 
akan dicari solusi dari masalah nilai nilai awal
x'' + 5x' + 5x = 0 , x(0) = 1 , x'(0) = 0
"""

# packages yang digunakan
from sympy import Function, dsolve, Derivative, solve, simplify, symbols, Eq, E, cos, sin
from sympy.abc import t
import numpy as np
from matplotlib import pyplot as plt

# penyelesaian
x = Function ('x')
sol = dsolve(Derivative(x(t), t, t) + 5*Derivative(x(t), t) + 5*x(t), x(t)).rhs
cnd0 = Eq(sol.subs(t, 0), 1)
cnd1 = Eq(sol.diff(t).subs(t,0), 0)

C1, C2 = symbols("C1, C2")
C1C2_sl = solve([cnd0, cnd1], (C1, C2))
sol1 = simplify(sol.subs(C1C2_sl))
print(Eq(x(t), sol1))

# Penggambaran Kurva Solusi 
print("\t\tSketsa Kurva Solusi")
t = np.linspace(0,20,1000)
def x(t) :
  return (((3*np.exp(2*t) - 1)*np.exp((-3*t))/2))
plt.plot(t,x(t))
plt.xlabel('t')
plt.ylabel('x(t)')
plt.xlim([0,6])
plt.ylim([-2,1.5])
plt.grid()
plt.show() 
````

Persamaan diferensial x'' + 5x' + 5x = 0 dengan x(0) = 1 dan x'(0) = 0 memiliki solusi x(t) = ((1 - sqrt(5))*exp(t*(5 - sqrt(5))/2) + (1 + sqrt(5))*exp(t*(sqrt(5) + 5)/2))*exp(-5*t)/2).

````
# Soal Latihan Nomor 2 
# mx'' + cx' + kx = F(t), x(0) = x_0, x'(0) = v_0
"""
Nomor 2
Untuk m = 1 c = 0, k = 4, x_0 = 0, dan x'(0) = 0, sketsa solusi persamaan diferensial tersebut 
untuk F(t) = cos(2t)

Penyelesaian: 
akan dicari solusi dari masalah nilai nilai awal
x'' + 4x = cos(2t) , x(0) = 0 , x'(0) = 0
"""

# packages yang digunakan
from sympy import Function, dsolve, Derivative, solve, simplify, symbols, Eq, E, cos, sin
from sympy.abc import t
import numpy as np
from matplotlib import pyplot as plt

# penyelesaian 
x = Function ('x')
sol = dsolve(Derivative(x(t), t, t) + 4*x(t) - cos(2*t), x(t)).rhs
cnd0 = Eq(sol.subs(t, 0), 0)
cnd1 = Eq(sol.diff(t).subs(t,0), 0)

C1, C2 = symbols("C1, C2")
C1C2_sl = solve([cnd0, cnd1], (C1, C2))
sol1 = simplify(sol.subs(C1C2_sl))
print(Eq(x(t), sol1))

# Penggambaran Kurva Solusi 
print("\t\tSketsa Kurva Solusi")
t = np.linspace(0,20,1000)
def f(t) :
  return ((t*np.sin(2*t))/4)
plt.plot(t,f(t),t,+t/4,t,-t/4)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.xlim([0,20])
plt.ylim([-10,10])
plt.grid()
plt.show() 
````

Persamaan diferensial x'' + 4x = cos(2t) dengan x(0) = 0 dan x'(0) = 0 memiliki solusi x(t) = (t/4*sin(2/t)).
Kurva solusi persamaan ini mengalami resonansi yang dibatasi oleh kurva amplop x(t) = +(t/4) dan -(t/4)
