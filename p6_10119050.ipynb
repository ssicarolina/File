````
# SOAL LATIHAN
# menggambar bidang fasa untuk sistem dengan A = [[-2,0],[0,-1]] dan
# beberapa kurva solusi dengan syarat awal x(0) = 0 dan y(0) = -10, -9, ... , 10
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
#menggambar bidang fasa
x = np.linspace(-10,10,25)
y = np.linspace(-10,10,25)
x,y = np.meshgrid(x,y)
plt.quiver(x,y, -2*x,-y)
plt.xlabel('$x$')
plt.ylabel('$y$')

#menggambar beberapa kurva solusi
def model(U,t):
    x,y = U
    dxdt = -2*x
    dydt = -y
    return [dxdt,dydt]
for i in range (-10,11):
    tspan = np.linspace(0, 50, 200)
    z0 = [-5,i]
    ys = odeint(model, z0, tspan)
    plt.plot(ys[:, 0],ys[:, 1]);
    z0 = [5,i]
    ys = odeint(model, z0, tspan)
    plt.plot(ys[:, 0],ys[:, 1]);
plt.xlim([-10,10])
plt.ylim([-10,10])
plt.show()



````

````
# SOAL LATIHAN
# NOMOR 2
# menggambar bidang fasa untuk sistem dengan A = [[-3,1],[2,-2]] dan
# beberapa kurva solusi dengan syarat awal x(0) = 0 dan y(0) = -10, -9, ... , 10
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
#menggambar bidang fasa
x = np.linspace(-10,10,25)
y = np.linspace(-10,10,25)
x,y = np.meshgrid(x,y)
plt.quiver(x,y, -3*x + y,2*x - 2*y)
plt.xlabel('$x$')
plt.ylabel('$y$')

#menggambar beberapa kurva solusi
def model(U,t):
    x,y = U
    dxdt = -3*x + y
    dydt = 2*x - 2*y
    return [dxdt,dydt]
for i in range (-10,11):
    tspan = np.linspace(0, 50, 2000)
    z0 = [0.1,i]
    ys = odeint(model, z0, tspan)
    plt.plot(ys[:, 0],ys[:, 1]);
plt.xlim([-10,10])
plt.ylim([-10,10])
plt.show()

# Penentuan jenis kestabilan
A = np.array([[-3,1],[2,-2]])
x0 =np.array([[0],[0]])

# evaluate the eigen values 
eigval, eigvec = np.linalg.eig(A)
lambda1 = eigval[0]
lambda2 = eigval[1]
print("\tNILAI EIGEN")
print("λ1: ",lambda1)
print("λ2: ",lambda2)

print("\nYaitu λ1 < λ2 < 0 , diperoleh bidang fasa berbentuk node(simpul) stabil asimtotik")
````

````
# SOAL LATIHAN
# NOMOR 3
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
#menggambar bidang fasa
x = np.linspace(-10,10,25)
y = np.linspace(-5,5,25)
x,y = np.meshgrid(x,y)
plt.quiver(x,y, x*(1 - x - y), y*(0.75 - y - 0.5*x))
plt.xlabel('$x$')
plt.ylabel('$y$')
 
#menggambar beberapa kurva solusi
def model(U,t):
    x,y = U
    dxdt = x*(1 - x - y)
    dydt = y*(0.75 - y - 0.5*x)
    return [dxdt,dydt]
for i in range (-3,10):
    tspan = np.linspace(0, 50, 2000)
    z0 = [10,i]
    ys = odeint(model, z0, tspan)
    plt.plot(ys[:, 0],ys[:, 1]);
plt.xlim([-10,10])
plt.ylim([-5,5])
plt.show()

print("Titik kesetimbangan berada di x,y yaitu (0,0), (1,0), (0.5, 0.5), (0, 0.75)")
````

Pada titik kesetimbangan (0,0) bidang fasa berbentuk (Im)Proper Node Tidak Stabil 

Pada titik kesetimbangan (1,0) bidang fasa berbentuk saddle

Pada titik kesetimbangan (0.5,0.5) bidang fasa berbentuk (Im)Proper Node Stabil Asimtotik

Pada titik kesetimbangan (0,0.75) bidang fasa berbentuk saddle

````
# Soal Tes Nomor 2 
# menggambar bidang fasa untuk sistem dengan A = [[7,1],[-4,3]] dan
# beberapa kurva solusi dengan syarat awal x(0) = 0 dan y(0) = -10, -9, ... , 10
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
#menggambar bidang fasa
x = np.linspace(-10,10,25)
y = np.linspace(-10,10,25)
x,y = np.meshgrid(x,y)
plt.quiver(x,y,7*x + y,-4*x + 3*y)
plt.xlabel('$x$')
plt.ylabel('$y$')

#menggambar beberapa kurva solusi
def model(U,t):
    x,y = U
    dxdt = 7*x + y
    dydt = -4*x + 3*y
    return [dxdt,dydt]
for i in range (-10,11):
    tspan = np.linspace(0, 50, 2000)
    z0 = [0,i]
    ys = odeint(model, z0, tspan)
    plt.plot(ys[:, 0],ys[:, 1]);
plt.xlim([-10,10])
plt.ylim([-10,10])
plt.show()

# Penentuan jenis kestabilan
A = np.array([[7,1],[-4,3]])
x0 =np.array([[0],[0]])

# evaluate the eigen values 
eigval, eigvec = np.linalg.eig(A)
lambda1 = eigval[0]
lambda2 = eigval[1]
print("λ1: ",lambda1)
print("λ2: ",lambda2)

print("\nλ1 = λ2 > 0, maka diperoleh bidang fasa dengan titik kesetimbangan di (0,0)berbentuk (Im)Proper Node tidak stabil)")
````

````
# SOAL Tes Nomor 3
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
#menggambar bidang fasa
x = np.linspace(-5,5,25)
y = np.linspace(-5,5,25)
x,y = np.meshgrid(x,y)
plt.quiver(x,y, x*(1.5 - x - 0.5*y), y*(2 - y - 0.75*x))
plt.xlabel('$x$')
plt.ylabel('$y$')
 
#menggambar beberapa kurva solusi
def model(U,t):
    x,y = U
    dxdt = x*(1.5 - x - 0.5*y)
    dydt = y*(2 - y - 0.75*x)
    return [dxdt,dydt]
for i in range (-11,10):
    tspan = np.linspace(0, 50, 2000)
    z0 = [1,i]
    ys = odeint(model, z0, tspan)
    plt.plot(ys[:, 0],ys[:, 1]);
plt.xlim([-5,5])
plt.ylim([-5,5])
plt.show()

print("Titik kesetimbangan berada di x,y yaitu (0,0),(1.5 0), (0.8, 1.4) dan (0,2)")
````
