````
# Soal Latihan Nomor 2 
# packages yang digunakan 
import numpy as np
from sympy import *


# Symbolic definition
t,C1, C2 = symbols("t C1 C2")
x,y = symbols ("x y", cls = Function, Function= True)

A = np.array([[2,5],[0,1]])
x0 =np.array([[4],[7]])

# evaluate the eigen values 
eigval, eigvec = np.linalg.eig(A)
lambda1 = eigval[0]
lambda2 = eigval[1]
x1 = eigvec[:,0].reshape(2,1)
x2 = eigvec[:,1].reshape(2,1)

# Fundamental Matrix
Psi = Matrix([[exp(lambda1*t)*np.float(x1[0]), exp(lambda2*t)*np.float(x2[0])],[exp(lambda1*t)*np.float(x1[1]), exp(lambda2*t)*np.float(x2[1])]])
Psi0 = Psi.subs(t,0)
Psi0 = np.array([[Psi0[0,0], Psi0[0,1]], [Psi0[1,0], Psi0[1,1]]], dtype='float')
soln = np.matmul(np.matmul(Psi,np.linalg.inv(Psi0)), x0)
print(soln)
#tunjukkin matriks fundamentalnya -HM
````
