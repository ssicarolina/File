```
import matplotlib.pyplot as plt
distance = float(input("Masukkan jarak dari jembatan ke tanah = ")) # 80
projected_area = float(input("Masukkan luas daerah proyeksi = ")) # 0.1
unweighted_length = float(input("Masukkan panjang kabel sebenarnya = ")) # 30
mass = float(input("Masukkan massa bungee jumper = ")) # 60 atau 70 atau 80
velocity = 0
acceleration_due_to_gravity = 9.81
weight = mass * acceleration_due_to_gravity
simulation_length = float(input("Masukkan panjang simulasi = ")) # 20
delta_t = float(input("Masukkan delta t = ")) # 0.02
numIterations = simulation_length/delta_t
spring_constant = 0
ground = True
while ground:
  print(spring_constant)
  velocity = 0
  time = 0
  length = 0
  length_max = distance
  l = []
  t = []
  l.append(length)
  t.append(time)
  positive = True
  for i in range(1,int(numIterations)+1):
    if length > unweighted_length:
      restoring_spring_force = -spring_constant*(length - unweighted_length)
    else:
      restoring_spring_force = 0
    air_friction = -0.65 * projected_area * velocity * abs(velocity)

    total_force = weight + restoring_spring_force + air_friction

    acceleration = total_force / mass
    velocity = velocity + acceleration*delta_t
    if positive:
      if velocity < 0:
        positive = False
        length_max = length
    length = length + velocity*delta_t
    time = i * delta_t
    l.append(length)
    t.append(time)
  if length_max < distance:
    ground = False
  spring_constant = spring_constant + 1
plt.plot(t,l)
plt.show()
```
