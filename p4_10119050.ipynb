````
# Soal Latihan Nomor 3
# packages yang digunakan 
from sympy.integrals.transforms import laplace_transform
from sympy.integrals.transforms import inverse_laplace_transform
from sympy import *
from sympy.abc import t, x, s, X, f, F
y, f, Y = symbols('y f Y', cls = Function)
t = symbols('t', positive = True)
s = symbols('s', real = True)


f = 1 - Heaviside(t-(3*pi))
y0 = 1
v0 = 0
#input hasil transformasi Laplace persamaan diferensial
print("\t\tf(t):\n")
eom = Eq((s**2*Y(s)-s*y0-v0) + 4*(s*Y(s)-y0), laplace_transform(f, t, s, noconds = True))

Yofs = solve(eom,Y(s))
soln = inverse_laplace_transform(Yofs[0],s,t)
soln_simp = expand(soln)
soln_simp
````

````
# Soal  Nomor 3
# packages yang digunakan 
from sympy.integrals.transforms import laplace_transform
from sympy.integrals.transforms import inverse_laplace_transform
from sympy import *
from sympy.abc import t, x, s, X, f, F
y, f, Y = symbols('y f Y', cls = Function)
t = symbols('t', positive = True)
s = symbols('s', real = True)


f = cos(t)
y0 = 1
v0 = 0
#input hasil transformasi Laplace persamaan diferensial
print("\t\tf(t):\n")
eom = Eq((s**2*Y(s)-s*y0-v0) + 2*(s*Y(s)-y0) + 2*Y(s), laplace_transform(f, t, s, noconds = True))

Yofs = solve(eom,Y(s))
soln = inverse_laplace_transform(Yofs[0],s,t)
soln_simp = expand(soln)
soln_simp
````
