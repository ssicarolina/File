```
# 3 menggambar medan gradien persamaan diferensial dan kurva solusinya
#b
from sympy import Function, dsolve, Derivative, symbols, lambdify
from sympy.abc import x
import numpy as np
import math
from matplotlib import pyplot as plt
x = np.linspace(-2.5,2.5,25)
y = np.linspace(-2.5,2.5,25)

def diff(x,y):                      # persamaan diferensial
  return y/(1+x**2)                         # turunan d/dy (x)

for j in x:
  for k in y :
    slope = diff(j,k)
    domain = np.linspace (j-0.02,j+0.02,2)
    def fun(x1,y1):
      z = slope*(domain-x1)+y1
      return z
    plt.plot(domain,fun(j,k),'k',
             solid_capstyle='projecting',
             solid_joinstyle='bevel')
    
y = Function('y')
x = symbols('x')

sol = dsolve(Derivative(y(x),x) - y(x)/(1+x**2), y(x)).rhs          # Solusi Umum 
print("Solusi Umum: ",sol)

c = np.linspace(-10,10,20)                                                  # y(0) = -0.9,-0.8, ... ,0
for i in range (len(c)):
    constants = solve ([sol.subs(x,0) - c[i]])                              # Mencari nilai c1 untuk setiap masalah nilai awal, y(0) = c
    final_answers = sol.subs(constants)                                     # Solusi khusus untuk setiap masalah nilai awal, y(0) = c
    y = lambdify(x, final_answers, modules=['numpy'])                      # Nilai y diperoleh dengan mensubtitusikan nilai x ke solusi 'final_answers'
    x_val = np.linspace(-2.5, 2.5, 100)                                     # Plot x dari -2.5 sampai 2.5
    y_val = y(x_val)                                                        # Dicari nilai y_val untuk setiap nilai x_val
    plt.plot(x_val, y_val, label ='y(0) = c = '+str('%.1f' % c[i]))         # Plot ditampilkan
        
plt.xlabel("x")
plt.ylabel("f (x)")
plt.legend(bbox_to_anchor=(1, 1.0275))
plt.xlim(-2.5,2.5)
plt.ylim(-2.5,2.5)
plt.show()

```
