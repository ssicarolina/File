````
# Nomor 1
import numpy as np
import sympy
from sympy import Function, dsolve, solve, Derivative, symbols
from sympy.abc import x
y = Function('y')
sol_1 = dsolve(Derivative(y(x),x) + y(x) - sympy.exp(-x),y(x)).rhs            # Solusi Umum
constants_1 = solve([sol_1.subs(x,-1)- 0.0])                # substitusi
final_answer_1 = sol_1.subs(constants_1)                    # solusi khusus
print("Solusi Umum No.1:" ,sol_1)
print("Nilai Konstanta no.1:", constants_1)
print("Solusi Akhir no.1:", final_answer_1)
````

````
# Nomor 2
# mencari solusi tunggal persamaan
from sympy import Function, dsolve, Derivative, symbols
from sympy.abc import x
y = Function('y')
sol = dsolve(Derivative(y(x),x) - y(x) - x - 1, y(x)).rhs          # Solusi Umum 
konstanta = solve([sol.subs(x,0) - 1]) # Mempunyai solusi yang tunggal
final = sol.subs(konstanta)
print("Solusi Umum No.2: ",sol)
print("Nilai Konstanta No.2:",konstanta)
print("Solusi No. 2:",final)
````

````
# Nomor 4
import numpy as np
from matplotlib import pyplot as plt
x = np.linspace (-5,5,25)
y = np.linspace(-2,5,25)

# Persamaan Diferensial
def diff(x,y):
  return y*(y-2)*(y-4)

# menggambar medan gradien
for j in x:
  for k in y :
    slope = diff(j,k)
    domain = np.linspace (j-0.02,j+0.02,2)
    def fun(x1,y1):
      z = slope*(domain-x1)+y1
      return z
    plt.plot(domain,fun(j,k),'k',
             solid_capstyle='projecting',
             solid_joinstyle='bevel')

plt.xlabel('x')
plt.ylabel('p(x)')
plt.xlim([-5,5])
plt.ylim([-2,5])
plt.show()
````

````
# Soal nomor 1
# menggambar medan gradien dan titik ekuilibriumnya

import numpy as np
from matplotlib import pyplot as plt
x = np.linspace (-6,6,25)
y = np.linspace(-6,6,25)

# Persamaan Diferensial
def diff(x,y):
  return -y*(y-2)*(y-4)/10

# menggambar medan gradien
for j in x:
  for k in y :
    slope = diff(j,k)
    domain = np.linspace (j-0.02,j+0.02,2)
    def fun(x1,y1):
      z = slope*(domain-x1)+y1
      return z
    plt.plot(domain,fun(j,k),'k',
             solid_capstyle='projecting',
             solid_joinstyle='bevel')

plt.xlabel('x')
plt.ylabel('p(x)')
plt.xlim([-6,6])
plt.ylim([-6,6])
plt.show()
````

Dengan melihat medan gradien didapat titik ekuilibrium yaitu y=0 (stabil), y=2(tidak stabil), dan y=4 (stabil)

````
# soal nomor 2b
# memerika eksistensi dan ketunggaln solusi dari persamaan diferensial
import numpy as np
import sympy
from sympy import Function, dsolve, solve, Derivative, symbols
from sympy.abc import x
y = Function('y')
sol = dsolve(Derivative(y(x),x) - 2*y(x)/x ,y(x)).rhs            # Solusi Umum

# untuk a = -1
print("\t\ta = -1")
constants_1 = solve([sol.subs(x,-1)+ 1.0])                # substitusi
final_answer_1 = sol.subs(constants_1)                    # solusi khusus
print("Solusi Umum No.2a:" ,sol)
print("Nilai Konstanta no.2a:", constants_1)
print("Solusi Akhir no.2a:", final_answer_1)

# untuk a = 0
print("\n\n\t\ta = 0")
constants_2 = solve([sol.subs(x,0)- 0.0])                # substitusi
final_answer_2 = sol.subs(constants_2)                    # solusi khusus
print("Solusi Umum No.2a:" ,sol)
print("Nilai Konstanta no.2a:", constants_2)
print("Solusi Akhir no.2a:", final_answer_2)

# untuk a = 1
print("\n\n\t\ta = 1")
constants_3 = solve([sol.subs(x,1)- 1.0])                # substitusi
final_answer_3 = sol.subs(constants_3)                    # solusi khusus
print("Solusi Umum No.2a:" ,sol)
print("Nilai Konstanta no.2a:", constants_3)
print("Solusi Akhir no.2a:", final_answer_3)
````

Dari hasil uji untuk a = -1,0,1 didapat bahwa persamaan diferensial dy/dx = 2y/x dengan y(a) = a untuk a = -1 dan a = 1 solusi tunggal, sementara untuk a = 0 memiliki solusi yang tidak tunggal
