```
# Fungsi yang menguji password valid atau tidak
def uji(x): 
    b = True
    if len(x) != 8: 
        print('Password belum 8 karakter')   
        b = False 
    if any(char.isdigit() for char in x): 
        b = False   
    if not any(char.isupper() for char in x): 
        print('Password belum berisi hurus Kapital') 
        b = False    
    if not any(char.islower() for char in x): 
        print('Password belum berisi huruf kecil') 
        b = False    
    if b: 
        return b 
a = 1
# Main method 
while a <= 3:
  print("\t\t\tPercobaan ke-",a)
  password = input("\t\t\tMasukkan Password: ")
  a = a +1
  if (uji(password)): 
    print("\t\t\tSelamat, Password Anda sudah valid") 
    a = 5
  else:
    print("Akibatnya, Password tidak valid !!") 
if a==4:
  print("\t\t\tMaaf, Percobaan Anda sudah habis")
          
```
