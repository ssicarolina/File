## Kasus 1: Titik eq 1 stabil, titik eq 2 tidak eksis

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

#1 liter darah manusia mengandung 6* 10^9 fagosit
#anggap ukuran 1 sel manusia = 8 * 10 ^-9 liter, maka dalam sel tersebut ada sekitar 50 fagosit, jadi anggap f0 = 50
# fagosit umurnya 1-3 hari kan, anggap diambil usia maksimal (3 hari)

alpha_f = 20                                    # konstanta produksi alami fagosit
theta = 0.02                                     # rasio produksi fagosit akibat adanya patogen
beta_f = 1/3                                    # rasio kematian alami fagosit
alpha_p = 1/2                                   # rasio produksi alami patogen
beta_p = 0.46                                   # rasio kematian alami patogen
delta_p = alpha_p - beta_p                      # selisih antara rasio kelahiran dan kematian alami patogen 
mu = 0.001                                        # rasio kematian patogen akibat diserang fagosit

#menggambar bidang fasa
x = np.linspace(0,900,20)
y = np.linspace(-10,150,20)
x,y = np.meshgrid(x,y)
plt.quiver(x,y,alpha_f + theta*x*y - beta_f*x, delta_p*y - mu*y*x)
plt.xlabel('$x$')
plt.ylabel('$y$')
#menggambar beberapa kurva solusi
def model(U,t):
  x,y = U
  dxdt = alpha_f + theta*x*y - beta_f*x
  dydt = delta_p*y - mu*y*x
  return [dxdt,dydt]

#anggap ada 3 jenis nilai awal patogen (lebih kecil, sama, lebih besar dari fagosit awal)
for i in [30,50,70]:
  tspan = np.linspace(0, 200, 1000)
  z0 = [50,i]
  ys = odeint(model, z0, tspan)
  plt.plot(ys[:, 0],ys[:, 1]);

plt.xlim([0,900])
plt.ylim([-10,80])
plt.xlabel("fagosit")
plt.ylabel("patogen")
```

```python
#DIAGRAM BIFURKASI P-C
import matplotlib.pyplot as plt
import numpy as np

# parameter 
alpha_f = 20                                    # konstanta produksi alami fagosit
theta = 0.02                                     # rasio produksi fagosit akibat adanya patogen
beta_f = 1/3                                    # rasio kematian alami fagosit
alpha_p = 1/2                                   # rasio produksi alami patogen
beta_p = 0.46                                   # rasio kematian alami patogen
delta_p = alpha_p - beta_p                      # selisih antara rasio kelahiran dan kematian alami patogen 
mu = 0.001                                        # rasio kemati


def p(c):
  return (beta_f*(1+c)*delta_p - alpha_f*mu)/(delta_p*theta)

c = np.linspace(2/3, 3, 200)
c1 = np.linspace(0,2/3, 200)
dx = 0
p1 = [i*dx for i in range (200) ]

# plot diagram bifurkasi

plt.plot(c1,p1,label="Titik eq I")
plt.plot(c,p(c), label="Titik eq II")
plt.legend()
plt.xlabel("c")
plt.ylabel("p")
```

```python
#DIAGRAM BIFURKASI F-C
import matplotlib.pyplot as plt
import numpy as np

# parameter 
alpha_f = 20                                    # konstanta produksi alami fagosit
theta = 0.02                                     # rasio produksi fagosit akibat adanya patogen
beta_f = 1/3                                    # rasio kematian alami fagosit
alpha_p = 1/2                                   # rasio produksi alami patogen
beta_p = 0.46                                   # rasio kematian alami patogen
delta_p = alpha_p - beta_p                      # selisih antara rasio kelahiran dan kematian alami patogen 
mu = 0.001                                        # rasio kemati

def f1(c1):
  return alpha_f/(beta_f*(1+c1))

c2 = np.linspace(2/3, 3, 200)
c1 = np.linspace(0,2/3, 200)
dx = delta_p/mu
f2 = [dx for i in range (200)]

# plot diagram bifurkasi
plt.plot(c1,f1(c1),label="Titik eq I")
plt.plot(c2,f2,label="Titik eq II")
plt.legend()
plt.xlabel("c")
plt.ylabel("f")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import numba

@numba.jit(nopython=True, parallel=True)
def calc():
  # Parameters
  min_r = 0.67
  max_r = 3.65
  step_r = 0.001
  max_iterations = 10000
  skip_iterations = 500
  max_counter = int((max_iterations - skip_iterations) * (max_r - min_r) / step_r)
  # The x and r results will be stored in these two arrays
  result_x = np.zeros(max_counter)
  result_r = np.zeros(max_counter)
  # Start the main loop
  i = 0
  for r in np.arange(min_r, max_r, step_r):
    x = 0.1
    for it in range(max_iterations):
      x = r * x * (1-x)
      if it > skip_iterations:
        result_x[i] = x
        result_r[i] = r
        i += 1
  result_x = result_x[result_r != 0].copy()
  result_r = result_r[result_r != 0].copy()
  return result_x, result_r

result_x, result_r = calc()

# Plot
plt.figure(figsize=(5, 3), dpi=200)
plt.plot(result_r, result_x, ",", color='k')
plt.show()
```
