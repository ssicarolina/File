## 1 (Jarak antar Anggota)

Diberikan sebuah vektor huruf kapital (vektor yang isinya huruf-huruf) dengan $n$ elemen (dimasukkan pengguna). Buat program Python untuk mencetak angka-angka dengan aturan berikut:

* Pada setiap huruf yang
berbeda dari vektor tersebut dicetak angka nol. Bila ada dua/lebih huruf yang sama maka huruf
kedua/lebih tersebut diganti dengan salah satu digit 1,2,...,9 yang menyatakan jarak/posisi
huruf tersebut terhadap huruf sebelumnya yang sama.

*Diasumsikan jarak antara dua huruf yang sama tidak melebihi 9.


Contoh masukan:
⃗$ \vec{a} = (’A’, ’A’, ’B’, ’C’, ’D’, ’B’, ’E’, ’F’, ’F’, ’E’, ’A’, ’B’, ’G’, ’B’, ’W’, ’B’)$

Keluaran: 0100030013960202

Penjelasan (tidak perlu di-print): Angka 1 di entri kedua ada karena di entri pertama sudah ada huruf A. Entri kelima berisikan B karena entri sebelumnya yang mengandung B adalah entri kedua.




````
# Jawaban nomor 1
# -*- coding: utf-8 -*-
"""
Created on Sun Sep 21 

@author: Carolina
"""
# Program mengisi vektor yang berukuran n entri yang isi datanya huruf kapital
# Program Menentukan dan menampilkan Jarak Antar Anggota dari vektor yang diketahui

# KAMUS 
#   a  : sebagai vektor dengan ukuran n,  yang berupa data string yaitu huruf kapital
#   b  : sebagai vektor dengan ukuran n, untuk menyimpan jarak antar anggota dari vektor a
#   n  : berupa data integer, sebagai variabel meyimpan nilai dari ukuran vektor a dan b


# ALGORITMA
# inisiasi panjang vektor a dan b
n = int(input("Masukkan jumlah suku vektor: "))
a = [0 for i in range(n)]
b = [0 for i in range (n)]
b[0] = 0

# Mengisi dan Menampilkan vektor a
for i in range(n):
    a[i] = str(input("Masukkan entri ke" +str(i+1)+" : "))

# Algoritma menentukan isi vektor b 
for i in range (n):
    j = i-1
    sama = 0
    while (j>=0) and sama==0:
        if a[i]==a[j]:
            sama = sama + 1
        j = j - 1
     
    if sama == 0:
        b[i] = 0
    else:
        b[i] = i - j -1
# Menampilkan vektor b
for i in range (n):
    print(b[i],end=" ")
````

#2 (Matriks Bilangan)

Diberikan sebuah matriks $A$ ukuran $5\times 5$. Gunakan perintah **randint** dari Python untuk mengisi elemen-elemen matriks tersebut dengan bilangan acak dalam rentang 10 - 100. Selanjutnya buat program Python untuk menukarkan isi baris ke 2 dengan baris ke 4. Setelah itu
tukarkan isi kolom ke 1 dengan kolom ke 3.

````
# Jawaban nomor 2
# -*- coding: utf-8 -*-
"""
Created on Sun Sep 21 

@author: Carolina
"""


# Program Mengisi Matriks 5x5 dengan nilai acak dari bilangan 10 sampai 100;
# Program menukar isi baris ke 2 dengan 4, kemudian menukar isi kolom ke 1 dengan 3, dari matriks yang sudah diketahui

# KAMUS 
#   A : Matriks 5x5 dengan nilai acak dari bilangan 10 sampai 100
#   NBrs, NKol :  data integer (ukuran brs & kol) 
#   i, j : int (indeks)
#   B : Matriks yang didapat  setelah baris ke 2 dengan 4 dari matriks A ditukar
#   K : Matriks yang didapat  setelah kolom ke 1 dengan 3 dari matriks B ditukar

# ALGORITMA

import random 

# deklarasi matriks A
NBrs = 5; NKol = 5;    
A = [[0 for j in range(NKol)] for i in range(NBrs)]

# Mengisi matriks A ukuran NBrsxNKol 
for j in range (NKol):
    for i in range (NBrs):
        A[i][j] = random.randint(10,100)

print("Matriks A")
# Menuliskan isi matriks ke layar 
for i in range (NBrs):
    # Cetak baris ke-i
    for j in range (NKol):
        print(str(A[i][j]) + " ", end='')
    # Baris ke-i selesai dicetak
    print()  # ganti baris

# Dekralasi matriks B dan K
B = [[0 for j in range(NBrs)] for i in range(NKol)]
K = [[0 for j in range(NBrs)] for i in range(NKol)]

# Inisiasi Matriks B dan K
for j in range (NKol):
    for i in range (NBrs):
        B[i][j] = A[i][j]
        K[i][j] = A[i][j]

# Algoritma mnedapatkan matriks B dari matriks A
print("Hasil dari isi baris ke 2 dengan baris ke 4 dari Matriks A ditukar yaitu ")
for i in range(NBrs):
    for j in range (NKol):
        if i !=1 and i !=3 :
            B[i][j] = A[i][j]
        if i == 1:
            B[i][j]= A[3][j]
        if i == 3:
            B[i][j] = A[1][j]
    for j in range (NBrs):
        if i != 1 and i !=3:
            B[i][j] = A[i][j]
        if i == 1:
            B[i][j] = A[3][j]
        if i == 3:
            B[i][j]=A[1][j]

# menampilkan matriks B
for i in range(NBrs):
    for j in range (NKol):
        print(B[i][j],end=" ")
    print()
    
# Algoritma mendapatkan matriks K dari matriks B
for i in range (NBrs):
    for j in range (NKol):
        if j != 0 and j !=2:
            K[i][j] = B[i][j]
        if j == 0:
            K[i][j] = B[i][2]
        if j == 2:
            K[i][j] = B[i][0]

# menampilkan matriks K
print("Hasil dari isi baris ke 2 dengan baris ke 4 ditukar") 
print("kemudian menukar isi Kolom ke 1 dengan kolom 3  Matriks A yaitu ")        
for i in range(NBrs):
    for j in range (NKol):
        print(K[i][j],end=" ")
    print()
    
````
