````

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square
from sympy import fourier_series, pi
from sympy.abc import t
from sympy import symbols
from sympy import *

f = (pi+t)*Heaviside(t+pi) - (pi+t)*Heaviside(t) + (pi-t)*Heaviside(t) - (pi-t)*Heaviside(t-pi) 
s = fourier_series(f, (t, -pi, pi))
s.truncate(5) # menampilkan 5 suku pertama deret fouriernya
deret = s.truncate(5)
print("Ekspresi simbolik Deret Fourier dengan 5 suku pertama fungsi yaitu: \n",deret)
````
