```
# Nomor 4a
import matplotlib.pyplot as plt
from scipy.integrate import odeint            #Ordinary Differential Equation
def logistic(y, t):                           # fungsi y' dalam y dan t
  """Returns the gradient dy/dt for the logistic equation"""
  return y*(y-2)*(y-4)
ts = np.linspace(-2, 2, 100) # membuat array dalam rentang waktu(t) dari -2 sampai 2 berukuran 100
x0 = 0.5 # an initial condition, x(0) = x0
xs = odeint(logistic, x0, ts)         #menyelesaikan persamaan differensial dengan x0 sebagai kondisi awal
# odeint returns an array of x values, at the times in ts.
plt.xlabel("$t$", fontsize=16)
plt.ylabel("$y$", fontsize=16)
plt.plot(ts, xs)

# Plot curves for several initial conditions
ics = np.linspace(0,4, 20) # a list of initial conditions
for x0 in ics:
  xs = odeint(logistic, x0, ts)
  plt.plot(ts, xs,"b")

plt.title("\nEquilibrium solutions of $y(y-2)(y-4)$")
plt.show()
```
