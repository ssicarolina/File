```python
# Metode Nilai Eigen
""" 
Contoh
Diberikan sitem dengan A = [[1 4][3 1]] dan x_0 = [1 0]
"""

import numpy as np
import matplotlib.pyplot as plt
from sympy import init_printing
init_printing()
from sympy import Function, Indexed, Tuple, sqrt, dsolve, solve, Eq, Derivative, sin, cos, symbols
from sympy.abc import k, t
from sympy import solve, Poly, Eq, Function, exp
from sympy.abc import x, y, z, a, b, c
f = Function('f')
from sympy import Indexed, IndexedBase, Tuple, sqrt
from sympy import *
from sympy.abc import *
from sympy.plotting import plot
init_printing()
t, C1, C2 = symbols("t C1 C2")
x, y = symbols("x y", cls = Function, Function = True)
eq1 = Eq(Derivative(x(t),t), x(t) + 4*y(t))
eq2 = Eq(Derivative(y(t),t), 3*x(t) + y(t))
soln = dsolve((eq1, eq2))
constants = solve((soln[0].subs(t,0).subs(x(0),1), soln[1].subs(t,0).subs(y(0),0)),{C1,C2})
xsoln = expand(soln[0].rhs.subs(constants))
print(xsoln)
ysoln = soln[1].rhs.subs(constants)
print(ysoln)
```
