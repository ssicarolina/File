## Membuat Plot (Introduction)

Untuk membuat plot pada python, kita memerlukan library tambahan karena python tidak memiliki fungsi bawaan untuk membuat plot.

Library matplotlib menyediakan fitur tersebut. Sehingga, sebelum membuat plot, maka kita perlu memanggil library tersebut dengan cara:

```python
import matplotlib.pyplot as plt
```

Setelah memanggil library tersebut, untuk membuat plot dan menampilkannya, maka kita membutuhkan ```plt.plot(x, y)``` dan ```plt.show()```. Sehingga kode akan menjadi:

```python
import matplotlib.pyplot as plt

# Definisikan terlebih dahulu x dan y
# x = ...
# y = ...

plt.plot(x, y)
plt.show()
```

## Membuat Titik pada Plot

Untuk memplot **sebuah** titik, maka kita memerlukan lokasi $(x, y)$ dari titik tersebut.

Selain itu, kita membutuhkan ```'o'``` (pada plt.plot()) untuk memberitahu bahwa setiap titik yang ingin kita tampilkan, memiliki bentuk bulat terisi. Jika anda tidak menggunakan ```'o'```, maka tidak akan ada titik yang tertampil. Lebih jauh tanpa itu hanya ada kurva saja yang ditampilkan

Misalkan kita ingin memplot titik $(3, 5)$, maka:

````
import matplotlib.pyplot as plt

x = 3
y = 5

plt.plot(x, y, 'o')
plt.show()
````

Untuk memplot banyak titik $\{(x_1, y_1), (x_2, y_2), ..., (x_n, y_n)\}$, maka anda perlu untuk menuliskan posisi (x dan y) dari seluruh titik tersebut. Posisi tersebut disimpan didalam suatu array $x$ dan $y$, dengan $x = \{ x_1, x_2, ..., x_n \}$ dan $y = \{ y_1, y_2, ..., y_n \}$

Sebagai contoh, misalkan anda ingin menampilkan titik $(2, 0), (3, 2), (1, 4),$ dan $(0, -1.5)$, maka anda menuliskan kode:

````
import matplotlib.pyplot as plt

x = [2, 3, 1, 0]
y = [0, 2, 4, -1.5]

plt.plot(x, y, 'o')
plt.show()
````

## Membuat Plot Kurva Sederhana

### Pengantar

Misalkan kita telah memiliki array $x = [x_1, x_2, ..., x_n]$ dan $y = [y_1, y_2, ..., y_n]$. Jika kita menggunakan ```plot(x, y)```, maka matplotlib akan **menggambarkan garis dari $(x_1, y_1)$ ke $(x_2, y_2)$, lalu $(x_2, y_2)$ ke $(x_3, y_3)$**, dan seterusnya.

Sehingga, **urutan** dari titik-titik tersebut sangat penting.

Sebagai contoh, berikut adalah fungsi $y = 4x - x^2$. Akan dibandingkan pengaruh dari pengurutan titik $x = [0, 1, 2, 3, 4]$ dengan $x = [1, 3, 4, 2, 0]$

````
# x = [0, 1, 2, 3, 4]
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y = [0, 3, 4, 3, 0]    # y[i] = f(x[i]), i = {0, 1, ..., 4}
plt.plot(x, y)
plt.show()
````

````
# x = [1, 3, 4, 2, 0]
import matplotlib.pyplot as plt

x = [1, 3, 4, 2, 0]
y = [3, 3, 0, 4, 0]    # y[i] = f(x[i]), i = {0, 1, ..., 4}
plt.plot(x, y)
plt.show()
````

### Kurva $y = f(x)$

Secara umum untuk memplot suatu fungsi $y = f(x)$ maka kita memerlukan titik-titik $(x, y)$ atau $(x, f(x))$. Sehingga pada program, kita memerlukan:
- Array $x$ (berisikan $x_1, x_2, ..., x_n$, dengan nilai $x_1 < x_2 < ... < x_n$)
- Array $y$ (berisikan $y_1, y_2, ..., y_n$, dengan elemen $y$ pada urutan ke-i ($y_i$ atau $f(x_i)$) berkorespondensi dengan elemen $x$ pada urutan ke-i ($x_i$))
- Syntax ```plt.plot(x, y)``` dan ```plt.show()```


Sebagai contoh, berikut adalah cara manual untuk memplot fungsi $f(x) = x^2$ dengan $x = [0, 4]$ dan lebar partisi $1$.

````
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]    # y[i] = f(x[i]), i = {0, 1, ..., 4}
plt.plot(x, y)
plt.show()
````

Dapat dilihat bahwa plot kurang terlihat mulus. Sehingga diperlukan cara lain untuk memplot fungsi tersebut.

Jika partisi yang digunakan lebih kecil, maka semakin banyak titik x dan y yang diperlukan. Secara manual, hal itu tentu sangat melelahkan. Dengan menggunakan ```for``` pada saat inisiasi array, maka hasil yang didapat akan serupa dengan sebelumnya, namun dengan banyak titik yang lebih banyak.

Berikut adalah cara yang lebih baik untuk memplot $f(x) = x^2$ dengan $x = [0, 4]$ dan lebar partisi $0.1$:

````
import matplotlib.pyplot as plt
import math

def f(x):
    return x**2

a = 0       # Batas x kiri
b = 4       # Batas x kanan

dx = 0.1    # Lebar partisi x
Nx = math.floor((b-a)/dx) + 1   # Banyak partisi x (ditambah 1 karena titik ujung inklusif/masuk)

x = [a + i*dx for i in range(Nx)]   # x = [0, 0.1, 0.2, ..., 3.9, 4]
y = [f(x[i]) for i in range(Nx)]    # y = [f(0), f(0.1), ..., f(4)]

plt.plot(x, y)
plt.show()
````

### Kurva Parametrik

Untuk memplot suatu kurva parametrik $(x(t), y(t))$ pada $t = [t_1, t_2, ..., t_n]$, maka kita memerlukan:
- Array $x$ (berisikan $x_1, x_2, ..., x_n$, dengan elemen $x$ pada urutan ke-i ($x_i$) berkorespondensi dengan elemen $t$ pada urutan ke-i ($t_i$))
- Array $y$ (berisikan $y_1, y_2, ..., y_n$, dengan elemen $y$ pada urutan ke-i ($y_i$) berkorespondensi dengan elemen $t$ pada urutan ke-i ($t_i$))
- Syntax ```plt.plot(x, y)``` dan ```plt.show()```

Sebagai contoh, misal ingin dibuat kurva setengah lingkaran kanan menggunakan persamaan $(cos(\frac{t}{2}), sin(\frac{t}{2}))$, dengan domain $t = [-\pi, \pi]$ dan lebar partisi $dt = 0.1$. Maka kode-nya:

````
# Kurva parametrik setengah lingkaran kanan
import matplotlib.pyplot as plt
import math
from math import pi, sin, cos

# Batas domain t
a = -pi
b = pi

# Lebar dan banyak partisi t
dt = 0.1
Nt = math.floor((b - a)/dt) + 1     # (ditambah 1 karena titik ujung inklusif/masuk)

# Definisi array t, x, y
t = [a+(i*dt) for i in range(Nt)]   # t = [-pi, -pi+0.1, -pi+0.2, ..., pi]
def f(x):
    return cos(x/2)
def g(x):
    return sin(x/2)
x = [f(t[i]) for i in range(Nt)]
y = [g(t[i]) for i in range(Nt)]


# Plot
plt.plot(x, y)
plt.show()
````

## Plot dua fungsi/kurva atau lebih

Secara umum, jika kita ingin menampilkan dua atau lebih fungsi/kurva; sebagai contoh misalkan ingin ditampilkan $k$ buah persamaan $(x_{(1)}, y_{(1)}), (x_{(2)}, y_{(2)}), ..., (x_{(k)}, y_{(k)})$, maka kita memerlukan:
- Array $x_{(1)}$ dan $y_{(1)}$, masing-masing berukuran $n_1$
- Array $x_{(2)}$ dan $y_{(2)}$, masing-masing berukuran $n_2$
- ...
- Array $x_{(k)}$ dan $y_{(k)}$, masing-masing berukuran $n_k$

Banyak partisi dari satu fungsi dapat berbeda dari fungsi lain.

Terdapat dua cara untuk memplot dua atau lebih fungsi/kurva. Yang pertama adalah dengan menuliskan seluruh array $x_1, y_1, ..., x_k, y_k$ dalam satu ```plt.plot```:
```python
plt.plot(x1, y1, x2, y2, x3, y3)
plt.show()
```

Atau dengan menggunakan beberapa ```plt.plot``` **sebelum** menampilkan plot:
```python
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)
plt.show()
```

Sebagai contoh, akan kita tampilkan kurva $f(x) = sin(x)$ beserta $g(x) = cos(x)$, dengan domain $x$ = $[0, 2 \pi]$ dan lebar partisi $0.1$.

````
import matplotlib.pyplot as plt
import math
from math import pi

def f(x):
    return math.sin(x)

def g(x):
    return math.cos(x)


a = 0       # Batas x kiri
b = 2*pi    # Batas x kanan

dx = 0.1    # Lebar partisi x
Nx = math.floor((b-a)/dx) + 1   # Banyak partisi x (ditambah 1 karena titik ujung inklusif/masuk)

# Karena partisi sama, maka cukup ada 1 array x saja
x = [a + i*dx for i in range(Nx)]   # x = [0, 0.1, 0.2, ..., 2 pi]
y1 = [f(x[i]) for i in range(Nx)]    # y = [f(0), f(0.1), ..., f(2 pi)]
y2 = [g(x[i]) for i in range(Nx)]    # y = [g(0), g(0.1), ..., g(2 pi)]

plt.plot(x, y1, x, y2)
plt.show()

# Alternatif lain:
# plt.plot(x, y1)
# plt.plot(x, y2)
# plt.show()
````

Contoh lain adalah menampilkan kurva $\frac{dy}{dt} = t$ dengan $y(0) = 0$, yang dihitung menggunakan metode forward Euler dengan domain $t = [0, 4]$ serta selang partisi $1$; beserta kurva solusinya $y(t) = \frac{1}{2} t^2$ dengan domain $t = [0, 4]$ dan selang partisi $0.1$.

Perhatikan bahwa banyak titik diskrit antara kurva pertama dan kedua berbeda. Sehingga harus ada dua array $x$ (dalam kasus ini $t$) yang berbeda.

````
import matplotlib.pyplot as plt
import math

def f(x):
    return (x**2)/2


a = 0       # Batas t kiri
b = 4       # Batas t kanan

dt1 = 1    # Lebar partisi t
dt2 = 0.1
Nt1 = math.floor((b-a)/dt1) + 1   # Banyak partisi t (ditambah 1 karena titik ujung inklusif/masuk)
Nt2 = math.floor((b-a)/dt2) + 1   # Banyak partisi t (ditambah 1 karena titik ujung inklusif/masuk)

# Karena partisi t berbeda, maka harus ada 2 array t
t1 = [a + i*dt1 for i in range(Nt1)]    # t = [0, 0.5, 1.0, ..., 4]
t2 = [a + i*dt2 for i in range(Nt2)]    # t = [0, 0.1, 0.2, ..., 4]
y1 = [0 for i in range(Nt1)]             # y = [f(0), f(0.5), ..., f(4)]
y2 = [f(t2[i]) for i in range(Nt2)]       # y = [g(0), g(0.1), ..., g(4)]

for i in range(1, Nt1):
    y1[i] = y1[i - 1] + dt1*t1[i - 1]

plt.plot(t1, y1, t2, y2)
plt.show()

# Alternatif lain:
# plt.plot(t1, y1)
# plt.plot(t2, y2)
# plt.show()
````

## Mengubah batas sumbu x dan y yang ditampilkan pada plot

Secara otomatis, matplotlib akan memplot data dari $x$ dan $y$ terkecil sampai terbesar. Namun terkadang, kita ingin menampilkan dalam skala yang lain.

Untuk menampilkan data dengan menggunakan batas x dan y yang lain, kita dapat menggunakan ```plt.xlim(left, right)``` dengan ```plt.ylim(bottom, top)```. Secara struktur, kedua kode tersebut dapat berada dalam urutan apapun, namun keduanya harus berada sebelum ```plt.show()```. Variable ```left, rigft, bottom, top``` menyatakan secara terurut batas kiri x, kanan x, bawah y, dan atas y.

Berikut adalah contoh struktur plot dengan mengubah daerah plot:
```python
plt.plot(x1, y1, x2, y2, ...)
plt.xlim(left, right)
plt.ylim(bottom, top)
plt.show()
```

Sebagai ilustrasi. Misalkan akan diplot fungsi $f(x) = x, g(x) = x^2,$ dan $h(x) = x^2 - x$, untuk domain $x = [0, 2]$ dengan selang partisi $0.1$ untuk seluruh fungsi. Daerah akan ditampilkan dengan batas $x$ dari $0$ hingga $2$ dan $y$ dari $0$ hingga $4.5$

````
import matplotlib.pyplot as plt
import math

def f(x):
    return x

def g(x):
    return x**2

def h(x):
    return x**2 - x

a = 0       # Batas x kiri
b = 2       # Batas x kanan

dx = 0.1    # Lebar partisi x
Nx = math.floor((b-a)/dx) + 1   # Banyak partisi x (ditambah 1 karena titik ujung inklusif/masuk)

# Karena partisi sama, maka cukup ada 1 array x saja
x = [a + i*dx for i in range(Nx)]   # x = [0, 0.1, 0.2, ..., 2]
y1 = [f(x[i]) for i in range(Nx)]   # y = [f(0), f(0.1), ..., f(2)]
y2 = [g(x[i]) for i in range(Nx)]   # y = [g(0), g(0.1), ..., g(2)]
y3 = [h(x[i]) for i in range(Nx)]   # y = [h(0), h(0.1), ..., h(2)]

plt.plot(x, y1, x, y2, x, y3)
##### xlim dan ylim #####
plt.xlim(0, 2)
plt.ylim(0, 4.5)
plt.show()
````

## Mengubah warna dan bentuk kurva

### Mengubah warna

Untuk mengubah warna pada kurva, terdapat tiga cara yang akan dibahas pada praktikum ini, yakni:
- Menggunakan nama warna tersebut secara full (contoh ```plot(x, y, color='Red')```). Silahkan cek https://matplotlib.org/3.1.0/gallery/color/named_colors.html untuk daftar yang lebih lengkap.
- Menggunakan string hex warna (contoh ```plot(x, y, '#FF0000')```)
- Menggunakan inisial warna (contoh ```plot(x, y, 'r')```)

Berikut adalah beberapa daftar warna. Jika anda ingin menggunakan warna lain yang tidak terdaftar pada tabel di bawah, maka silahkan gunakan hex atau cek website di atas.

Jika terdapat beberapa kurva yang ingin diwarnai, maka sebaiknya anda menggunakan ```plt.plot(x, y)``` secara terpisah untuk setiap kurva, karena akan lebih mudah untuk dipahami.

```python
plt.plot(x1, y1, 'r')
```

Sebagai contoh, misalkan ingin ditampilkan dua fungsi $f(x) = x^2 - x + 0.25, g(x) = -x^2 + x,$ dan $h(x) = 0.25$ pada domain $x = [0, 1]$ dengan selang partisi $0.05$. Ketiga fungsi tersebut akan memiliki warna hijau, merah, dan oranye secara berurutan.

````
import matplotlib.pyplot as plt
import math

def f(x):
    return x**2 - x + 0.25

def g(x):
    return -x**2 + x

a = 0       # Batas x kiri
b = 1       # Batas x kanan

dx = 0.05    # Lebar partisi x
Nx = math.floor((b-a)/dx) + 1   # Banyak partisi x (ditambah 1 karena titik ujung inklusif/masuk)

# Karena partisi sama, maka cukup ada 1 array x saja
x = [a + i*dx for i in range(Nx)]   # x = [0, 0.1, 0.2, ..., 2]
y1 = [f(x[i]) for i in range(Nx)]   # y = [f(0), f(0.1), ..., f(2)]
y2 = [g(x[i]) for i in range(Nx)]   # y = [g(0), g(0.1), ..., g(2)]
y3 = [0.25 for i in range(Nx)]   # y = [h(0), h(0.1), ..., h(2)]

plt.plot(x, y1, "g")        # Hijau
plt.plot(x, y2, color='#FF0000')  # Merah
plt.plot(x, y3, color="Orange")   # Oranye
plt.show()
````

Sebagai contoh, misalkan ingin ditampilkan dua fungsi $f(x) = x^2 - x + 0.25, g(x) = -x^2 + x,$ dan $h(x) = 0.25$ pada domain $x = [0, 1]$ dengan selang partisi $0.05$. Ketiga fungsi tersebut akan memiliki bentuk:
- $f(x)$ bulatan tanpa garis, dengan warna hijau
- $g(x)$ garis patah-patah dengan warna merah
- $h(x)$ bulatan dengan garis, dengan oranye

````
import matplotlib.pyplot as plt
import math

def f(x):
    return x**2 - x + 0.25

def g(x):
    return -x**2 + x

a = 0       # Batas x kiri
b = 1       # Batas x kanan

dx = 0.05    # Lebar partisi x
Nx = math.floor((b-a)/dx) + 1   # Banyak partisi x (ditambah 1 karena titik ujung inklusif/masuk)

# Karena partisi sama, maka cukup ada 1 array x saja
x = [a + i*dx for i in range(Nx)]   # x = [0, 0.1, 0.2, ..., 2]
y1 = [f(x[i]) for i in range(Nx)]   # y = [f(0), f(0.1), ..., f(2)]
y2 = [g(x[i]) for i in range(Nx)]   # y = [g(0), g(0.1), ..., g(2)]
y3 = [0.25 for i in range(Nx)]   # y = [h(0), h(0.1), ..., h(2)]

plt.plot(x, y1, "og")
plt.plot(x, y2, "--r")
plt.plot(x, y3, "o-", color='#ED7B00')
plt.show()
````

## Judul plot dan penamaan sumbu

### Judul plot



Untuk memberikan judul plot, anda dapat menggunakan ```plt.title()```. Seperti biasa, kode tersebut anda ketik sebelum ```plt.show()```.

Secara struktur:
```python
plt.plot(x1, y1, x2, y2)
plt.title("JUDUL")

```

Sebagai contoh, kita akan menampilkan kurva $f(x) = x$ untuk domain $x = [0, 1]$ dengan selang partisi $0.01$. Judul plot ini dimisalkan "Kurva Sederhana"

````
import matplotlib.pyplot as plt
import math

def f(x):
    return x

a = 0       # Batas x kiri
b = 1       # Batas x kanan

dx = 0.01    # Lebar partisi x
Nx = math.floor((b-a)/dx) + 1   # Banyak partisi x (ditambah 1 karena titik ujung inklusif/masuk)

# Karena partisi sama, maka cukup ada 1 array x saja
x = [a + i*dx for i in range(Nx)]   # x = [0, 0.1, 0.2, ..., 2]
y = [f(x[i]) for i in range(Nx)]   # y = [f(0), f(0.1), ..., f(2)]

plt.plot(x, y)
plt.title("Kurva Sederhana")
plt.show()
````

### Penamaan sumbu

Untuk memberi nama pada sumbu-x dan y adalah dengan menggunakan ```plt.xlabel()``` dan ```plt.ylabel()```, secara berurutan. Seperti biasa, kode tersebut diletakkan sebelum ```plt.show()```.

Secara struktur:
```python
plt.plot(x1, y1, x2, y2)
plt.xlabel("Sumbu x")
```

Sebagai contoh, kita akan menampilkan kurva $f(x) = x$ untuk domain $x = [0, 1]$ dengan selang partisi $0.01$. Sumbu-x dan y akan diberi nama "Massa (g)" serta "Konsentrasi (g/ml)" secara berurutan.

````
import matplotlib.pyplot as plt
import math

def f(x):
    return x

a = 0       # Batas x kiri
b = 1       # Batas x kanan

dx = 0.01    # Lebar partisi x
Nx = math.floor((b-a)/dx) + 1   # Banyak partisi x (ditambah 1 karena titik ujung inklusif/masuk)

# Karena partisi sama, maka cukup ada 1 array x saja
x = [a + i*dx for i in range(Nx)]   # x = [0, 0.1, 0.2, ..., 2]
y = [f(x[i]) for i in range(Nx)]   # y = [f(0), f(0.1), ..., f(2)]

plt.plot(x, y)
plt.xlabel("Massa (g)")
plt.ylabel("Konsentrasi (g/ml)")
plt.show()
````

## Grid pada plot

Untuk membuat plot menjadi terlihat lebih jelas, maka kita dapat menggunakan grid pada plot. Secara simpel, grid akan menampilkan kotak-kotak didalam plot untuk menunjukkan nilai x dan y tertentu.

Menampilkan grid cukup dengan menambahkan ```plt.grid()``` sebelum ```plt.show()```. Sehingga, secara struktur:
```python
plt.plot(x1, y1, x2, y2)
plt.grid()
plt.show()
```

Sebagai contoh, kita akan menampilkan kurva $f(x) = x^2$ untuk domain $x = [0, 3]$ dengan selang partisi $0.01$.

````
import matplotlib.pyplot as plt
import math

def f(x):
    return x**2

a = 0       # Batas x kiri
b = 3       # Batas x kanan

dx = 0.01    # Lebar partisi x
Nx = math.floor((b-a)/dx) + 1   # Banyak partisi x (ditambah 1 karena titik ujung inklusif/masuk)

# Karena partisi sama, maka cukup ada 1 array x saja
x = [a + i*dx for i in range(Nx)]   # x = [0, 0.1, 0.2, ..., 2]
y = [f(x[i]) for i in range(Nx)]   # y = [f(0), f(0.1), ..., f(2)]

plt.plot(x, y)
plt.grid()
plt.show()
````

## Legenda

Tentu untuk membuat pembaca lebih mengerti plot yang kita buat, maka kita memerlukan legenda.

Untuk membuat legenda, anda dapat menggunakan ```plt.legend()```. Namun kedua kurva perlu diberi label terlebih dahulu. Sebaiknya pendefinisian ```plt.plot()``` setiap kurva dipisah agar memudahkan anda. Sehingga, secara struktur:
```python
plt.plot(x1, y1, label="Kurva1")
plt.plot(x2, y2, label="Kurva2")
plt.legend()
plt.show()
```

Sebagai contoh, akan kita tampilkan kurva $f(x) = sin(x)$ beserta $g(x) = cos(x)$, dengan domain $x$ = $[0, 2 \pi]$ dan lebar partisi $0.1$. Sebagai tambahan, akan disediakan legenda untuk kedua fungsi tersebut.

````
import matplotlib.pyplot as plt
import math
from math import pi

def f(x):
    return math.sin(x)

def g(x):
    return math.cos(x)


a = 0       # Batas x kiri
b = 2*pi    # Batas x kanan

dx = 0.1    # Lebar partisi x
Nx = math.floor((b-a)/dx) + 1   # Banyak partisi x (ditambah 1 karena titik ujung inklusif/masuk)

# Karena partisi sama, maka cukup ada 1 array x saja
x = [a + i*dx for i in range(Nx)]   # x = [0, 0.1, 0.2, ..., 2 pi]
y1 = [f(x[i]) for i in range(Nx)]    # y = [f(0), f(0.1), ..., f(2 pi)]
y2 = [g(x[i]) for i in range(Nx)]    # y = [g(0), g(0.1), ..., g(2 pi)]

plt.plot(x, y1, label="f(x) = sin(x)")
plt.plot(x, y2, label="g(x) = cos(x)")
plt.legend()
plt.show()
````

## Catatan

Penggunaan istilah (seperti ```'b'```, ```'m'```) yang tertuang pada bagian warna, style, dan marker, ditaruh sebelum parameter-parameter lain (label, warna). Sebagai contoh:
```python
plt.plot(x1, y1, "--r", label="Merah patah-patah")
plt.plot(x2, y2, "o", color="Orange", label="Oranye bulat")
```

Sebaiknya ```plt.plot()``` ditaruh di bagian paling awal dan ```plt.show()``` di bagian paling akhir. Sedangkan untuk yang lain (```plt.legend()```, ```plt.xlabel()```, ```plt.title()```, dsb) urutannya dibebaskan

# 2. Heatmap / Grid Plot

Cellular Automata salah satu dalam studi yang dianalisis pada dunia komputasi. Anda akan mempelajari materi ini pada minggu-minggu akhir perkuliahan.

Nantinya, pada bagian cellular automata, data akan disimpan dalam bentuk 2D (matriks), yakni berupa grid. Untuk menampilkan informasi tersebut, dibutuhkan fitur heatmap dari library ```seaborn```.

Sehingga kita akan mempelajari:
- Inisiasi data (matriks)
- Pembuatan heatmap/grid plot

## Inisiasi data

Karena data yang akan digunakan merupakan data 2D, maka kita membutuhkan matriks untuk menyimpan data tersebut. Namun, karena dari awal kita menggunakan array bawaan dari python, maka kita akan lanjut menggunakan array bawaan tersebut untuk membangun matriks.

Misalkan ingin dibangun matriks:
```
0 1 2
0 2 1
1 1 1
```

Maka, anda dapat membuat array didalam array seperti di bawah:
```
A = [[0, 1, 2],
     [0, 2, 1],
     [1, 1, 1]]
```

Dapat kita bayangkan bahwa baris pertama matriks disimpan sebagai elemen pertama A, baris kedua matriks disimpan sebagai elemen kedua A, dst.

Lebih lanjut, jika kita ingin mendefinisikan matriks berukuran $m \times n$ (m buah baris, n buah kolom) yang elemen-elemennya bernilai nol, maka:
```python
A = [[0 for j in range(n)] for i in range(m)]
```

Perhatikan bahwa looping j akan membuat sebuah array berukuran $1 \times n$, lalu array tersebut akan dibuat sebanyak $m$ kali oleh looping i. Sehingga kita dapatkan matriks berukuran $m \times n$ yang elemen-elemennya bernilai nol.

---

Untuk mengakses elemen pada matriks A, maka kita dapat memanggil dengan cara ```A[i][j]```, dengan i menyatakan indeks baris yang ingin diakses dan j menyatakan indeks kolom yang akan diakses. Seperti biasa, i dan j dimulai dari 0.

---

Sebagai ilustrasi, akan diberikan contoh membangun sebuah matriks dengan aturan tertentu.

Misalkan kita ingin membangun matriks berukuran $6 \times 6$ dengan aturan nilai elemennya sama dengan jumlah dari indeks baris dan kolom elemen tersebut. Maka:

````
A = [[0 for j in range(6)] for i in range(6)]

# Inisiasi matriks dengan aturan sebelumnya
for i in range(6):
    for j in range(6):
        A[i][j] = i + j     # Mengakses elemen pada matriks A



# Print matriks
print("A = [", end="")
for i in range(6):
    print(A[i], end="")
    if i == 5:
        print("]")
        break
    else:
        print()
    print("     ", end="")
````

## Pembuatan Heatmap / Grid Plot

Seperti yang telah disebutkan sebelumnya, untuk menampilkan nilai pada suatu matriks menjadi informasi dalam bentuk warna pada suatu grid/sel, kita membutuhkan library ```seaborn```.

Sehingga pada baris pertama/awal program, disisipkan:
```python
import seaborn as sns
```
```as sns``` dicantumkan agar kita dapat memanggil library tersebut secara lebih singkat, yakni cukup dengan ```sns``` saja.

---
Untuk membuat grid/sel berwarna tersebut, kita dapat memanggil dan memasukkan grid tersebut kedalam suatu variable (sebutlah ax):
```
ax = sns.heatmap(data)
```

Atau, jika ingin dispesifik-an set warna grid: (untuk warna yang dapat dicek pada gambar yang ada di website https://matplotlib.org/3.3.2/tutorials/colors/colormaps.html#miscellaneous, sebagai contoh warna: 'viridis', 'plasma', dll)
```
ax = sns.heatmap(data, cmap=...)
```

```data``` pada kedua baris program diatas merupakan matriks.

---

Sebagai contoh, akan ditampilkan pola papan catur yang berukuran $8 \times 8$.

Jika anda perhatikan, kotak pada papan catur akan berwarna putih (1) ketika jumlahan indeks kolom dan baris dari kotak tersebut bernilai genap (karena indeks python dimulai dari 0); dan akan berwarna hitam (0) ketika jumlahan indeks kolom dan baris bernilai ganjil. Sehingga:

````
import seaborn as sns

# Inisiasi matriks data, berdasarkan aturan di atas
data = [[0 for j in range(8)] for i in range(8)]
for i in range(8):
    for j in range(8):
        if (i + j) % 2 == 0:   # Genap
            data[i][j] = 1
        else:                # Ganjil
            data[i][j] = 0

# Plot heatmap
ax = sns.heatmap(data)
````

## Catatan

Sebagai catatan, struktur dari membuat heatmap atau plot grid adalah:
- Bangun matriks
- Plot menggunakan ```sns.heatmap()```

### Membuat array

Sebelum kita dapat menggunakan array, tentunya kita harus menyiapkan array tersebut. Terdapat beberapa cara untuk menyiapkan array.

Cara pertama adalah dengan menulis seluruh anggota yang diinginkan secara manual.

````
# Membuat array
A0a = [0, 1, -1, 5]
A0b = []
A0c = ["Hello", "Guys"]

# Menampilkan array
print("A0a = ", A0a)
print("A0b = ", A0b)
print("A0c = ", A0c, end='\n\n')
````

Cara lain (dan nantinya cara ini lebih sering digunakan) adalah dengan menggunakan looping for, baik didalam maupun diluar inisiasi array tersebut.

````
# Inisiasi array dengan looping di dalam []
A1 = [0 for i in range(4)]          # A1 = [0, 0, 0, 0]
A2 = [1 for i in range(6)]          # A2 = [1, 1, 1, 1, 1, 1]
A3 = [i for i in range(5)]          # A3 = [0, 1, 2, 3, 4]
A4 = [i/3 for i in range(4)]        # A4 = [0, 1/3, 2/3, 3/3] = [0, 0.33, 0.66, 1]
A5 = [i**2 + 1 for i in range(4)]   # A5 = [1, 2, 5, 10, 17]

def f(x):
    return x**2 - 2
A6 = [f(i) for i in range(5)]  # A6 = [f(0), f(1), f(2), f(3), f(4)]

# Menampilkan array
print("A1 = ", A1)
print("A2 = ", A2)
print("A3 = ", A3)
print("A4 = ", A4)
print("A5 = ", A5)
print("A6 = ", A6)
````

````
# Inisiasi array dengan looping di luar []
A7 = []
A8 = []
for i in range(6):
    A7.append(i**3)
    A8 = A8 + [-i]

print("A7 = ", A7)
print("A8 = ", A8)
````

### Menggabungkan array

Untuk menggabungkan dua atau lebih array yang berbeda, kita dapat menggunakan operator ```+```. Sebagai contoh:

````
A = [1, 2, 3]
B = [-1, -2, -3]

C = A + B
D = B + A

print("C = ", C)
print("D = ", D)
````

Cara ini pula dapat digunakan untuk mendefinisikan suatu array, dengan menggabungkan array yang mau kita definisikan dengan array berisikan 1 elemen secara berulang. Sebagai contoh:

````
A = []

def f(x):
    return x**2

for i in range(5):
    A = A + [f(i)]

print("A = ", A)
````

### Slicing array

Slicing array adalah mengambil elemen dari suatu array pada indeks awal hingga suatu indeks akhir.

Secara lengkap cara menggunakan fungsi ini adalah ```A[start:stop:step]```, dengan ```start``` adalah indeks mulai pengambilan, ```stop``` adalah indeks akhir penghentian, serta ```step``` adalah penambahan indeks (dalam kata lain, selisih antara dua indeks yang bersebelahan).

Sebagai catatan, python **tidak mengambil elemen pada indeks ```stop```**, melainkan akan berhenti mengambil saat menyentuh atau melewati indeks ```stop ``` tersebut.

Secara default, jika kita memberitahu nilai:
- ```start```, maka diasumsikan bernilai 0
- ```stop```, maka diasumsikan bernilai panjang array tersebut
- ```step```, maka diasumsikan bernilai 1

Sebagai contoh:
```
A[::2] = A[0, 2, 4, 6, ..., Nx - 1 (atau Nx - 2)]
A[1:5] = A[1, 2, 3, 4]
A[2::] = A[2, 3, 4, ..., Nx - 1]
A[:4:] = A[0, 1, 2, 3]
A[:6:2] = A[0, 2, 4]
```

````
A = [i for i in range(8)]   # A = [0, 1, 2, ..., 7]
B = A[1:6]

print(A)
print(B)
````

### Duplikasi array

Misalkan pada suatu hari anda ingin menduplikat array ```A``` ke array ```B``` . Secara naif, anda dapat menggunakan ```A = B```. Namun perhatikan kode di bawah:

````
A = [1, 2, 3, 4]
B = A               # Cara yang salah untuk menduplikat array

# Hanya B saja yang ditambahkan elemennya
B.append(5)

print("A = ", A)
print("B = ", B)
````

Terlihat bahwa dengan menggunakan ```A = B```, kedua array masih terikat. Dalam artian, jika salah satu array berubah elemennya, maka array lain juga akan berubah. Tentu kita tidak menginginkan hal ini terjadi. Sehingga dibutuhkan cara lain untuk menduplikat array.

Ada dua cara untuk menduplikat array.

Yang pertama adalah dengan menduplikat setiap elemen pada array pertama ke array kedua. Sebagai contoh:

````
A = [1, 2, 3, 4]
n = len(A)
B = [A[i] for i in range(n)]    # Duplikasi array A ke B dengan duplikat setiap elemen

B.append(5)

print("A = ", A)
print("B = ", B)
````

Terlihat dari cara diatas, bahwa array A dan B berbeda ketika ditampilkan setelah array B ditambahkan suatu elemen.

Cara lain duplikasi array adalah dengan menggunakan ```.copy()```. Sebagai contoh:

````
A = [1, 2, 3, 4]
B = A.copy()        # Duplikasi array A ke B dengan .copy

B.append(5)

print("A = ", A)
print("B = ", B)
````

Terlihat hasil yang serupa antara cara ```.copy()``` dengan cara sebelumnya.
