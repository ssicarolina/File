```python
# Nomor 2a

# f(t) = merupakan fungsi ganjil
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square
from sympy import fourier_series, pi
from sympy.abc import t
from sympy import symbols
from sympy import *
f = (-t)*Heaviside(t+2) - (-t)*Heaviside(t) + (t)*Heaviside(t) - (t)*Heaviside(t-2)
s = fourier_series(f, (t, -2, 2))
s.truncate(40) # menampilkan 10 suku pertama deret 

p = plot(f, s.truncate(40), (t, -3*pi, 3*pi), show = False, legend = True)
p[0].line_color = 'red'
p[0].label = 'f(t)'
p[1].line_color = 'blue'
p[1].label = 'n = 40'
p.show()
```

```python
from sympy import *
from sympy.abc import t
from sympy.utilities.lambdify import lambdify
import numpy as np
import matplotlib.pyplot as plt

f = Heaviside(t)
s = fourier_series(f, (t, -pi, pi)) #simbolik
k= 10
fun_fourier = lambdify(t, s.truncate(k)) # simbolik
x = np.linspace(-np.pi, np.pi, 100) # numerik
plt.plot(x, np.heaviside(x, 0.5), linewidth = 2)
plt.plot(x, fun_fourier(x), linewidth=2)
plt.grid()
plt.legend(['f(t)=t**2', 'Deret Fourier (10 suku pertama)'])

plt.show()
```
