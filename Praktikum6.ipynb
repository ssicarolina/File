Permainannya seperti bowling, namun hanya ada satu pin dan lintasan bolanya tidak mendatar, melainkan bisa dibentuk sesuka hati (walaupun harus berupa turunan agar bola dapat melaju tanpa gaya dorong). Pemenang permainan ini adalah yang bolanya berhasil mengenai pin lebih dahulu.

Kurva **brakistokron** merupakan contoh sikloid yang dapat digambarkan dengan persamaan parametrik 
 \begin{array}{l}
x( \theta ) \ =\ R( \theta -sin( \theta ))\\
y( \theta ) \ =\ R( 1-cos( \theta ))
\end{array}

Dengan  y menyatakan ketinggian dengan sumbu positif searah gravitasi, x menyatakan jarak horizontal dengan sumbu positif mengarah ke pin, R adalah jari-jari pembentuk sikloid dan θ merupakan sudut antara arah gravitasi dan arah gerak bola.

Misal posisi awal bola di (0,0) dan pin berada di (1,0.3). Beberapa asumsi yang digunakan adalah gravitasi tetap di setiap titik, serta tidak ada gesekan sehingga berlaku hukum kekekalan energi. Gunakan g = 10.

Diketahui nilai θ di titik awal adalah 0 dan nilai θ di posisi pin, θ2 dapat dihitung secara numerik dengan menyalin kode berikut di python :

```
# This is formatted as code
from scipy.optimize import newton
def f(theta):
    return y2/x2 - (1-np.cos(theta))/(theta-np.sin(theta))
theta2 = newton(f, np.pi/2)
```

selanjutnya, hitung R dengan mensubstitusi nilai θ2 yang didapat ke persamaan parametrik.

````python
# PROGRAM soal1a
# pertama program akan menentukan nilai theta2 yang dibutuhkan
# untuk menentukan posisi akhir bola bowling pada persamaan
# parametrik. Selanjutnya program juga akan menentukan nilai R.
# Terakhir program akan menampilkan plot lintasan bowling milik Saya

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton


# bagian ini akan mencari solusi numerik f(theta) = 0
# dengan kata lain, theta yang memenuhi persamaan
#   y(theta)/x(theta) = y2/x2
# dan menyimpannya di variabel `theta2`. Karena dari
# dekripsi soal `theta2` menandakan posisi pin, kita
# simpulkan `x2` dan `y2` perlu diisi dengan koordinat pin
x2, y2 = 1, 0.3
f = lambda theta: y2/x2 - (1-np.cos(theta))/(theta-np.sin(theta))
theta2 = newton(f, np.pi/2)

# dari persamaan parametrik, x(theta2) = x2 = 1, didapatkan
R = 1/(theta2 - np.sin(theta2))


# selanjutnya kita tinggal membuat visualisasi brakistokron
# buat array 100 elemen yang evenly-spaced dengan ujung kiri
# interval 0 dan ujung kanan interval theta2
Theta = np.linspace(0, theta2, 100)

# lalu, buat array pergerakan bowling pada interval Theta
#   untuk nilai Y, kita perlu mengambil negatif nilainya,
#   agar mengikuti orientasi yang diberikan soal
X = R*(Theta - np.sin(Theta))
Y = -R*(1 - np.cos(Theta))


# waktunya plotting
plt.figure(figsize=(8,4))   # atur ukuran plot
plt.grid(True)              # tambahkan grid
plt.title('Lintasan Bola Bowling Milik Saya')
plt.plot(X, Y);
````

````python
# PROGRAM soal1c
# program akan menentukan siapa yang memenangkan balapan(?) bowling
# program akan menggunakan nilai-nilai yang sudah didapatkan di
# bagian PROGRAM soal1a


g = 10      # konstanta gravitasi
m = y2/x2   # koreksi soal: m adalah gradien lintasan bowling milik Alien

# waktu tempuh lintasan brakistokron
Tbar = theta2 * np.sqrt(R/g)
# waktu tempuh lintasan linear
Tlin = np.sqrt((1+m**2)*x2/(g*m))

print('Saya' if Tbar<Tlin else 'Alien', 'memenangkan balapan.')
````

b. Warnai array $M$ yang telah kalian buat sehingga sel yang berbeda warna memiliki warna yang berbeda. Anda dibebaskan memakai warna background atau warna tulisan. (Bonus: Buatlah agar angka nol tidak ditampilkan)

Berikut adalah salah satu pendekatan untuk menyelesaikan permasalahan tersebut.

````python
# PROGRAM soal2a
# Untuk menyelesaikan permasalahan ini, program akan
# membentuk matriks nol ukuran 10x10, lalu mengisi
# koordinat yang sesuai dengan bom, dilanjutkan dengan
# menandai daerah sekitar bom dengan angka yang sesuai


import numpy as np


M = np.zeros((10, 10), dtype=int)   # buat matriks nol 10x10, isi integer
NIM = input('masukkan NIM: ')
NIM += NIM[0]                       # memudahkan ketika menaruh bom

# taruh bom
for i in range(8):
    M[int(NIM[i]), int(NIM[i+1])] = -1
    
# tandai daerah sekitar bom
#   jika (x,y) adalah lokasi bom, daerah yang perlu
#   ditandai adalah (x+dx,y+dy) dengan dx,dy bernilai
#   -1, 0, atau 1. Tentu, perlu dicek agar jangan
#   menandai lokasi bom atau lokasi diluar definisi
for x in range(10):
    for y in range(10):
        
        # jika (x,y) bukan bom, langsung
        # pergi ke iterasi berikutnya saja!
        if M[x,y]!=-1: continue
        
        # (x,y) bom, tandai daerah sekitarnya
        for dx in [-1,0,1]:
            for dy in [-1,0,1]:
                
                #cek apakah valid untuk menandai posisi
                if (0<=x+dx<=9) and (0<=y+dy<=9) and (M[x+dx, y+dy]!=-1):
                    # tambah nilai di posisi ini
                    # bahasa lain: "di posisi ini ada satu bom lagi"
                    M[x+dx, y+dy] += 1

# cetak matriks
print(M)
````

````python
# PROGRAM soal2b_metode1
# Membuat visualisasi dari matriks M
# yang dihasilkan PROGRAM soal2a


import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
plt.title('Heatmap Minesweeper')
plt.imshow(M);
````

````python
# PROGRAM soal2b_metode2
# Membuat visualisasi dari matriks M yang dihasilkan PROGRAM soal2a
#   Dapat dipahami bila matplotlib tidak mudah bagi pemula, karena
#   kita perlu 'berurusan' dengan hal-hal yang terasa 'low-level'.
#   Namun, dapat berurusan dengan hal-hal ini memberikan kebebasan
#   bagaimana sebuah plot ingin ditampilkan -- tapi mengorbankan
#   waktu mengetik :'((( Jika anda tidak ingin repot, ada beberapa
#   package visualisasi lain yang lebih 'friendly', seperti `seaborn`


import matplotlib.pyplot as plt
from matplotlib import colors


# warna minesweeper :D
#   -1 = Black = Bom!
#    0 = White = Tidak ada bom di sekitar saya
#    1 = Cyan  = ada satu bom di sekitar saya
#    dst...
minecol = ['black', 'white', 'cyan', 'green', 'red',
           'blue', 'brown', 'teal', 'purple', 'grey']


# Menghasilkan objek `figure` dan objek `axes`. Jika anda membaca
# contoh dari internet, anda mungkin sering melihat kode berikut.
# Kode ini memudahkan kita dalam mengatur visualisasi:
#   `fig` berisi hal yang berurusan dengan figure hasil plotting,
#    misal ingin menyimpan plot sebagai jpg, dan sebagainya.
#   `ax` berisi hal yang berurusan plotting itu sendiri, seperti
#    mengatur judul, warna dan bentuk garis, skala, dan lain-lain.
fig, ax = plt.subplots()
fig.set_size_inches((6, 5))                   # atur ukuran gambar


cmap = colors.ListedColormap(minecol)         # warna untuk heatmap
norm = colors.BoundaryNorm(range(-1, 10), 10) # mapping data dari M dan cmap
heatmap = ax.imshow(M, cmap=cmap, norm=norm)  # buat heatmap dari matriks M
ax.set_xticks([x+0.5 for x in range(10)])     # geser agar tick sumbu x ada di sisi kanan elemen
ax.set_yticks([x+0.5 for x in range(10)])     # geser agar tick sumbu y ada di sisi bawah elemen
ax.grid(True)                                 # buat grid
ax.set_xticklabels([])                        # sembunyikan label tick di sumbu x
ax.set_yticklabels([])                        # sembunyikan label tick di sumbu y
ax.set_title('Heatmap Minesweeper')           # set judul heatmap


cbar = fig.colorbar(heatmap)                  # buat plot baru untuk legenda warna heatmap
cbar.set_ticks([x+0.5 for x in range(-1,10)]) # geser agar tick colorbar ada di tengah elemen 
cbar.set_ticklabels(range(-1,9))              # perbaiki label tick :D
````
