```
"""
6.1 Transformasi Laplace dan Inversnya
Sebagai contoh, akan ditentukan transformasi Laplace dari fungsi f(t) = u2(t) = u(t − 2)
(fungsi tangga satuan).
"""
# Mencari Transformasi Laplace dari Fungsi f(t) = u2(t) = u(t − 2)
from sympy.integrals.transforms import laplace_transform
from sympy import symbols
from sympy import *
t = symbols('t', positive = True)
s = symbols('s', real = True)
f = Heaviside(t-2)
print("F(s)= ",laplace_transform(f, t, s))


# Mencari Invers transformasi laplace dari Fungsi F(s) = e^-2s/s
from sympy.integrals.transforms import inverse_laplace_transform
from sympy import symbols
from sympy import *
t = symbols('t', positive = True)
s = symbols('s', real = True)
#F = (s**2 - (2*(s**2)*exp(-2*s*pi)) - 1)/((s**2 - 1)*(s - 1))
F = (2*s - 2*s*exp(-2*pi*s) - s**3 - s) / (1-s**4)
print("Invers F(s)= ",expand(inverse_laplace_transform(F, s, t)))
```

```
"""
6.2 Solusi Persamaan Diferensial Biasa
Sebagai contoh, diberikan masalah nilai awal 
x'' + 10x'+ 21x = δ(t − 2), x(0) = 4, dan x'(0) = 7
"""

# Mencari Solusi dari x'' + 10x'+ 21x = δ(t − 2), x(0) = 4, dan x'(0) = 7
from sympy.integrals.transforms import laplace_transform
from sympy.integrals.transforms import inverse_laplace_transform
from sympy import *
from sympy.abc import t, x, s, X, f, F
x, f, X = symbols('x f X', cls = Function)
t = symbols('t', positive = True)
s = symbols('s', real = True)
f = DiracDelta(t-2)
x0 = 4
v0 = 7
#input hasil transformasi Laplace persamaan diferensial
eom = Eq((s**2*X(s)-s*x0-v0)+10*(s*X(s)-x0)+21*X(s), laplace_transform(f, t, s, noconds = True))

Xofs = solve(eom,X(s))
soln = inverse_laplace_transform(Xofs[0],s,t)
soln_simp = expand(soln)
soln_simp
```
