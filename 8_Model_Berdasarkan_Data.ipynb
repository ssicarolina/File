```
"""
Fungsi Linear y = mx + c
Fungsi Eksponensial
Fungsi Polinom
Fungsi Akar
Fungsi Logaritmik
Fungsi Trigonometri
Fungsi Periodik
"""

""" Diberikan data
x = [0.2 0.4 0.3 0.3]
y = [0.1 0.3 0.3 0.6]
"""

import matplotlib.pyplot as plt
import math
x = [0.2,0.4,0.3,0.3]
y = [0.1,0.3,0.3,0.6]


a = 0       # Batas x kiri
b = 1      # Batas x kanan
dx = 0.01    # Lebar partisi x
Nx = math.floor((b-a)/dx) + 1   # Banyak partisi x (ditambah 1 karena titik ujung inklusif/masuk)

#                                     Fungsi Linear
def f(x):
    return 0.025 + x
x1 = [a + i*dx for i in range(Nx)]   # x = [0, 0.1, 0.2, ..., 3.9, 4]
y1 = [f(x1[i]) for i in range(Nx)]    # y = [f(0), f(0.1), ..., f(4)]

#                                     Fungsi Kuadrat
def f(x):
    return x**2
x2 = [a + i*dx for i in range(Nx)]   # x = [0, 0.1, 0.2, ..., 3.9, 4]
y2 = [f(x2[i]) for i in range(Nx)]    # y = [f(0), f(0.1), ..., f(4)]

#                                     Fungsi Sinus-Cosinus
import math
def f(x):
    return math.sin(x)
def g(x):
    return math.cos(x)
x3 = [a + i*dx for i in range(Nx)]   # x = [0, 0.1, 0.2, ..., 3.9, 4]
y3 = [f(x3[i]) for i in range(Nx)]    # y = [f(0), f(0.1), ..., f(4)]
y4 = [g(x3[i]) for i in range(Nx)]    # y = [f(0), f(0.1), ..., f(4)]
plt.plot(x, y,"o")
plt.plot(x1,y1,label ="Linear" )       
plt.plot(x2,y2,label ="Kuadrat" ) 
plt.plot(x3,y3,label ="Sinus" )   
plt.plot(x3,y4,label ="Cosinus" )
plt.legend()
plt.show()
```
