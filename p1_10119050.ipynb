````
"""Soal latihan di Modul (Hal.10) nomor 1,2(b),3(a),3(c), dan Soal latihan di Modul Hal.14
no 1."""
# Latihan Soal halaman 10
# Nomor 1
import numpy as np
x = np.array([[1],[2]])                             # inisialisasi nilai vektor x
A = np.array([[1,1],[1,1]])                         # inisialisasi Nilai Matriks A
a = np.linalg.det(A)                                # Determinan Matriks A
b_1= A*x                                            # operasi pada soal b
b_2= (A*A)*x
c = A**3                                            # operasi pada soal c
eigvals,eigvecs = np.linalg.eig(c)     
print("Determinan A: ",a)                           # menampilkan nilai dari determinan matriks A
print("A*x: ",b_1)                                  # menampilkan solusi soal (b)
print(" A*A*x:",b_2)                        
print("Nilai Eigen", eigvals)                        # menampilkan Nilai Eigen dari matriks A**3
print("Vektor Eigen", eigvecs)                       # menampilkan Vektor Eigen dari matriks A**3

# Nomor 2(b)
import matplotlib.pyplot as plt
import math
from math import e
import numpy as np
def b(x):                                       # merumuskan fungsi pada soal (b)
  y = (math.e)**(-x)
  y = y*(np.sin(2*x) + 3*np.cos(x))
  return y
x = np.linspace(-10,10,1000)                    # vektor dengan panjang 1000 dimulai dari -10 hingga 10
plt.plot(x,b(x))                                # memplot grafik b(x)
plt.xlabel('x')                                 # sumbu-x berlabel x
plt.ylabel('y_b')                               # sumbu-y berlabel
plt.xlim(-11,11)                                # batas sumbu-x dari -11 sampai 11
plt.ylim(-170,170)                              # batas sumbu-y dari -170 sampai 170
plt.show()                                      # menampilkan grafik

# Nomor 3(a) dan 3 (c)
from sympy import *
x = symbols('x')
a = limit((x**3)/(math.e)**x,x,oo)                    # operasi pada soal (a)
c = diff(diff(diff((sin(2*x)/x),x),x),x)              # operasi pada soal (c)    
print("3a:",a)                                        # menampilkan hasil soal (a)
print("3c:",c)                                        # menampilkan hasil soal (c)

# Latihan Soal Halaman 14
# Nomor 1

from sympy import Function, dsolve, solve, Derivative, symbols
from sympy.abc import x
y = Function('y')
x = Function('x')
sol_1 = dsolve((x**3)*Derivative(y(x),x) + 4*x**2*y(x) - np.exp(-x),y(x))            # Solusi Umum
constants_1 = solve([sol_1.subs(x,-1)-0.0])                             # substitusi
final_answer_1 = sol_1.subs(constants_1)                               # solusi khusus
print(sol_1)
print(constants_1)
print(final_answer_1)                                                 # menampilkan hasil solusi
````

````
# Soal nomor 1
import numpy as np
A = [[0,1,2],[1,1,1],[2,1,0]]
B = [[1,0,1],[0,1,0],[1,0,1]]

# pertanyaan pertama nilai-nilai eigen A dan B
# nilai eigen A
eigvals,eigvecs = np.linalg.eig(A) 
print("Nilai Eigen", eigvals)                        # menampilkan Nilai Eigen dari matriks A  
# nilai eigen B
eigvals,eigvecs = np.linalg.eig(B) 
print("Nilai Eigen", eigvals)                        # menampilkan Nilai Eigen dari matriks B   

# pertanyaan kedua
I = [[1,0,0],[0,1,0],[0,0,1]]                       # menginisialisasi matriks identitas I
# operasi pada pertanyaan kedua
d = np.matmul(B,B)
d = np.matmul(A,B)
C = I - d
print(C)                                      # hasil dari pertanyaan kedua dalam variabel C

# pertanyaan ketiga
# nilai eigen C
eigvals,eigvecs = np.linalg.eig(C) 
print("Nilai - nilai Eigen C", eigvals)                        # menampilkan Nilai Eigen dari matriks C  

````
