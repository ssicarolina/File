# Meminimumkan
Akan dicari titik minimum dari 
f(x) = x1^2 + 5x2^2 + 25x3^2 - x1 - x2 - x3 


```python
import numpy as np
```



# Metode Algoritma Umum



```python
x=np.array([0,0,0]) #nilai awal
alpha=0.01  #coba kombinasi (0,0,0,0.1) , (0,0,0,0.11), (0,0,0,0.05), (1,1,1,0.1) , (1,1,1,0.01), (1,1,1,0.05)
nablaf=np.array([2*x[0]-1 , 10*x[1]-1 , 50*x[2]-1]) #menghitung nabla f(x) dengan x awal

print("x0= ",x)
print("nabla f(x0)= ",nablaf)
print("norm(nabla f(x0))= ",np.linalg.norm(nablaf))

error=0.00001
iter=1
maxiter=200


while  np.linalg.norm(nablaf)>error and iter<maxiter:
    x = np.array(x - (alpha*np.array(nablaf))) #memperbaharui x
    nablaf=np.array([2*x[0]-1, 10*x[1]-1, 50*x[2]-1]) #menghitung nabla f(x) yang baru dengan x baru
    if(iter%5==0):
        print("\n","x",iter,"= ",x)
        print("nabla f(x",iter,")= ",nablaf)
        print("norm(nabla f(xx",iter,")= ",np.linalg.norm(nablaf))
    iter=iter+1

print("\n","x",iter-1,"= ",x)
print("nabla f(x",iter-1,")= ",nablaf)
print("norm(nabla f(xx",iter-1,")= ",np.linalg.norm(nablaf))
```

# Metode Steepest Descent



```python
#Metode Steepest Decent

x=np.array([0,0,0]) #nilai awal
nablaf=np.array([2*x[0]-1 , 10*x[1]-1 , 50*x[2]-1]) #menghitung nabla f(x) dengan x awal
nabla2f=np.array([[2,0,0],[0,10,0],[0,0,50]]) #kebetulan konstan, tidak bergantung pada x
alpha = (np.transpose(nablaf)@nablaf)/(np.transpose(nablaf)@nabla2f@nablaf)

print("x0= ",x)
print("nabla f(x0)= ",nablaf)
print("norm(nabla f(x0))= ",np.linalg.norm(nablaf))
print("alpha = ", alpha)


error=0.00001
iter=1
maxiter=100

while  np.linalg.norm(nablaf)>error and iter<maxiter:
  x=x-alpha*nablaf              #memperbaharui x
  print("\n","x",iter,"= ",x)
  nablaf=np.array([2*x[0]-1 , 10*x[1]-1 , 50*x[2]-1]) #menghitung nabla f(x) dengan x awal
  print("nabla f(x",iter,")= ",nablaf)
  print("norm(nabla f(x",iter,")= ",np.linalg.norm(nablaf))
  alpha = (np.transpose(nablaf)@nablaf)/(np.transpose(nablaf)@nabla2f@nablaf)
  print("alpha =" ,alpha)
  iter=iter+1
```

# Metode Newton

```python
#Metode Newton

x=np.array([0,0,0]) #nilai awal
nablaf=np.array([2*x[0]-1 , 10*x[1]-1 , 50*x[2]-1]) #menghitung nabla f(x) dengan x awal
nabla2f=np.array([[2,0,0],[0,10,0],[0,0,50]]) #kebetulan konstan, tidak bergantung pada x

print("x0= ",x)
print("nabla f(x0)= ",nablaf)
print("norm(nabla f(x0))= ",np.linalg.norm(nablaf))

error=0.00001
iter=1
maxiter=100

while  np.linalg.norm(nablaf)>error and iter<maxiter:
  x=x-np.linalg.inv(nabla2f)@nablaf #memperbaharui x
  print("\n","x",iter,"= ",x)
  nablaf=np.array([2*x[0]-1 , 10*x[1]-1 , 50*x[2]-1]) #menghitung nabla f(x) dengan x awal
  nabla2f=np.array([[2,0,0],[0,10,0],[0,0,50]]) #kebetulan konstan, tidak bergantung pada x
  print("nabla f(x",iter,")= ",nablaf)
  print("norm(nabla f(x",iter,")= ",np.linalg.norm(nablaf))
  iter=iter+1
```

# Metode Conjugate Gradien

```python
# metode conjugate gradien

x=np.array([0,0,0]) #nilai awal
nablaf=np.array([2*x[0]-1 , 10*x[1]-1 , 50*x[2]-1]) #menghitung nabla f(x) dengan x awal
nabla2f=np.array([[2,0,0],[0,10,0],[0,0,50]])
d=-nablaf #menghitung d

print("x0= ",x)
print("norm(nabla f(x0))= ",np.linalg.norm(nablaf))


iter=1
maxiter=100
error=0.000001

while  np.linalg.norm(nablaf)>error and iter<maxiter:
  alpha=-d@nablaf/(d@nabla2f@d)
  x=x+alpha*d
  print("x",iter,"= ", x)
  nablaf=np.array([2*x[0]-1 , 10*x[1]-1 , 50*x[2]-1])
  print("norm(nabla f(x",iter,"))= ",np.linalg.norm(nablaf)) 
  d=-nablaf+( nabla2f@nablaf@d/(d@nabla2f@d)  )*d
  iter=iter+1
```
